# 二部グラフ判定

##問題：二部グラフ判定

頂点数 N、辺数 M の無向グラフが与えられます。

各頂点を色 0 または色 1 で塗り、すべての辺について両端の頂点を異なる色にできるとき、このグラフを二部グラフと呼びます。

再帰的な深さ優先探索を用いて、与えられたグラフが二部グラフか判定してください。

グラフは連結とは限りません。

入力
N M
a1 b1
a2 b2
...
aM bM
N：頂点数
M：辺数
頂点には 0 から N - 1 の番号が付いている
ai bi は、頂点 ai と頂点 bi の間に辺があることを表す
出力

二部グラフなら、

Yes

二部グラフでなければ、

No

と出力してください。

## 関数うめ

```python
def is_bipartite(graph):
    n = len(graph)


# 入力
n, m = map(int, input().split())

graph = [[] for _ in range(n)]

for _ in range(m):
    a, b = map(int, input().split())

    # 無向グラフなので両方向に追加
    graph[a].append(b)
    graph[b].append(a)

# 判定
if is_bipartite(graph):
    print("Yes")
else:
    print("No")
```

## 答え

```python
def is_bipartite(graph):
    n = len(graph)

    # -1: 未訪問、0と1: 頂点の色
    color = [-1] * n

    def dfs(v, current_color):
        # 頂点vを塗る
        color[v] = current_color

        for next_v in graph[v]:

            # まだ塗られていないなら、反対の色で探索する
            if color[next_v] == -1:
                if not dfs(next_v, 1 - current_color):
                    return False

            # 隣接頂点が同じ色なら二部グラフではない
            elif color[next_v] == current_color:
                return False

        return True

    # グラフが連結でない場合もあるため、全頂点を確認する
    for v in range(n):
        if color[v] == -1:
            if not dfs(v, 0):
                return False

    return True


# 入力
n, m = map(int, input().split())

graph = [[] for _ in range(n)]

for _ in range(m):
    a, b = map(int, input().split())

    # 無向グラフなので両方向に追加
    graph[a].append(b)
    graph[b].append(a)

# 判定
if is_bipartite(graph):
    print("Yes")
else:
    print("No")
```
