# 二部グラフ判定

##問題：二部グラフ判定

頂点数 N、辺数 M の無向グラフが与えられます。

## 答え

```python
def is_bipartite(graph):
    color = [0] * len(graph)  # 0: 未訪問、1/-1: 色

    def dfs(v):
        for u in graph[v]:
            if color[u] == 0:
                color[u] = -color[v]
                if not dfs(u):
                    return False
            elif color[u] == color[v]:
                return False
        return True

    for v in range(len(graph)):
        if color[v] == 0:
            color[v] = 1
            if not dfs(v):
                return False

    return True


# 入力
n, m = map(int, input().split())

graph = [[] for _ in range(n)]

for _ in range(m):
    a, b = map(int, input().split())

    # 無向グラフなので両方向に追加
    graph[a].append(b)
    graph[b].append(a)

# 判定
if is_bipartite(graph):
    print("Yes")
else:
    print("No")
```

# トポロジカルソートの再起での実装（大事だが未暗記）

```python
def dfs(v, order):
    if v.color == 1:       # 探索中の頂点に戻った
        return False       # 循環あり

    if v.color == 2:       # すでに処理済み
        return True

    v.color = 1

    for next in v.nexts:
        if not dfs(next, order):
            return False

    v.color = 2
    order.append(v)        # 帰りがけに追加
    return True


def topological_sort(nodes):
    order = []

    for v in nodes:
        if not dfs(v, order):
            return None    # 循環あり

    order.reverse()
    return order
```

# トポロジカルソートのKahsでの実装(大事だが未暗記)

```python
from collections import deque

def topological_sort(adj):
    n = len(adj)

    # 入次数を計算
    indeg = [0] * n
    for u in range(n):
        for v in range(n):
            if adj[u][v] == 1:
                indeg[v] += 1

    # 入次数0の頂点をキューに入れる
    queue = deque()
    for v in range(n):
        if indeg[v] == 0:
            queue.append(v)

    order = []
    while queue:
        u = queue.popleft()
        order.append(u)
        for v in range(n):
            if adj[u][v] == 1:
                indeg[v] -= 1
                if indeg[v] == 0:
                    queue.append(v)

    if len(order) < n:
        return None
    return order
```

```python
## 自分であなうめ
import sys

sys.setrecursionlimit(10**6)

# 頂点数
N = int(input())

# 隣接リスト
graph = [[] for _ in range(N)]

# 木の辺は N - 1 本
for _ in range(N - 1):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

depth = [0] * N
subtree_size = [0] * N


def dfs(v, parent, d):
  # ここに実装
  depth[v] = d

  subtree_size[v] = 1

  for next_v in graph[v]:
    if next_v == parent:
      continue

    dfs(next_v, v, d+1)

    subtree_size[v] += subtree_size[next_v]


# 頂点0を根にする
dfs(0, -1, 0)

for v in range(N):
    print(
        "vertex =", v,
        "depth =", depth[v],
        "subtree_size =", subtree_size[v]
    )
```
