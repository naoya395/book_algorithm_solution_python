## code 13.1 グラフ探索の基本形

頂点 s から探索を始め、見つけた頂点を seen で記録します。

ここでは deque をキューとして使うので幅優先探索の形になります。深さ優先探索にしたい場合は、同じ todo をスタックとして使う考え方になります。

```python
from collections import deque


def search(G: list[list[int]], s: int) -> None:
    # 頂点数を調べる
    N = len(G)

    # seen[v] は、頂点 v をすでに発見したかどうか
    seen = [False] * N

    # 次に調べる頂点を入れておくキュー
    todo = deque()

    # 始点 s を発見済みにして、キューに入れる
    seen[s] = True
    todo.append(s)

    # キューが空になるまで探索する
    while todo:
        v = todo.popleft()

        # v から進める頂点をすべて調べる
        for x in G[v]:
            if seen[x]:
                continue

            # 新しく見つけた頂点を記録して、あとで調べる
            seen[x] = True
            todo.append(x)
```

## code 13.2 DFS

再帰関数 dfs で、まだ訪問していない頂点へ進みます。

全頂点を始点候補にすることで、グラフが連結でない場合でもすべての連結成分を探索できます。元コードに出力はないので、この Python 版も出力しません。

```python
import sys


sys.setrecursionlimit(10 ** 7)

seen: list[bool] = []


def dfs(G: list[list[int]], v: int) -> None:
    # 頂点 v を訪問済みにする
    seen[v] = True

    # v から行ける頂点を順に調べる
    for next_v in G[v]:
        if seen[next_v]:
            continue
        dfs(G, next_v)


def main() -> None:
    global seen

    N, M = map(int, input().split())

    # 有向グラフを隣接リストで作る
    G = [[] for _ in range(N)]
    for _ in range(M):
        a, b = map(int, input().split())
        G[a].append(b)

    # まだ訪問していない頂点から DFS を始める
    seen = [False] * N
    for v in range(N):
        if seen[v]:
            continue
        dfs(G, v)


if __name__ == "__main__":
    main()
```

## code 13.3 BFS

幅優先探索で、始点 0 から各頂点への最短辺数を求めます。

未訪問の頂点は dist[v] = -1 としておき、新しく見つけたら dist[v] + 1 で距離を更新します。

```python
from collections import deque


def BFS(G: list[list[int]], s: int) -> list[int]:
    N = len(G)
    dist = [-1] * N
    que = deque()

    # 始点 s の距離は 0
    dist[s] = 0
    que.append(s)

    while que:
        v = que.popleft()

        for x in G[v]:
            # すでに距離が決まっている頂点は調べない
            if dist[x] != -1:
                continue

            dist[x] = dist[v] + 1
            que.append(x)

    return dist


def main() -> None:
    N, M = map(int, input().split())

    # 無向グラフなので両方向に辺を追加する
    G = [[] for _ in range(N)]
    for _ in range(M):
        a, b = map(int, input().split())
        G[a].append(b)
        G[b].append(a)

    dist = BFS(G, 0)

    for v in range(N):
        print(f"{v}: {dist[v]}")


if __name__ == "__main__":
    main()
```

## code 13.4 到達可能性の判定

頂点 s から DFS をして、最後に seen[t] を見ます。

seen[t] が True なら s から t へ到達できます。

```python
import sys


sys.setrecursionlimit(10 ** 7)

seen: list[bool] = []


def dfs(G: list[list[int]], v: int) -> None:
    seen[v] = True

    for next_v in G[v]:
        if seen[next_v]:
            continue
        dfs(G, next_v)


def main() -> None:
    global seen

    N, M, s, t = map(int, input().split())

    G = [[] for _ in range(N)]
    for _ in range(M):
        a, b = map(int, input().split())
        G[a].append(b)

    # s から到達できる頂点をすべて探索する
    seen = [False] * N
    dfs(G, s)

    if seen[t]:
        print("Yes")
    else:
        print("No")


if __name__ == "__main__":
    main()
```

## code 13.5 二部グラフ判定

隣り合う頂点を 0 と 1 の別の色で塗れるかを DFS で調べます。

すでに塗られた隣接頂点が自分と同じ色なら、二部グラフではありません。

```python
import sys


sys.setrecursionlimit(10 ** 7)

color: list[int] = []


def dfs(G: list[list[int]], v: int, cur: int = 0) -> bool:
    # 頂点 v を cur 色で塗る
    color[v] = cur

    for next_v in G[v]:
        if color[next_v] != -1:
            # 隣接頂点が同じ色なら失敗
            if color[next_v] == cur:
                return False
            continue

        # 隣接頂点は 1 - cur で塗る
        if not dfs(G, next_v, 1 - cur):
            return False

    return True


def main() -> None:
    global color

    N, M = map(int, input().split())

    G = [[] for _ in range(N)]
    for _ in range(M):
        a, b = map(int, input().split())
        G[a].append(b)
        G[b].append(a)

    color = [-1] * N
    is_bipartite = True
    for v in range(N):
        if color[v] != -1:
            continue
        if not dfs(G, v):
            is_bipartite = False

    if is_bipartite:
        print("Yes")
    else:
        print("No")


if __name__ == "__main__":
    main()
```

## code 13.6 トポロジカルソート

DFS の帰りがけに頂点を order へ追加します。

最後に order.reverse() すると、辺の向きに沿った順序が得られます。

```python
import sys


sys.setrecursionlimit(10 ** 7)

seen: list[bool] = []
order: list[int] = []


def rec(G: list[list[int]], v: int) -> None:
    seen[v] = True
    for next_v in G[v]:
        if seen[next_v]:
            continue
        rec(G, next_v)

    # 帰りがけ順に追加する
    order.append(v)


def main() -> None:
    global seen, order

    N, M = map(int, input().split())
    G = [[] for _ in range(N)]
    for _ in range(M):
        a, b = map(int, input().split())
        G[a].append(b)

    seen = [False] * N
    order = []
    for v in range(N):
        if seen[v]:
            continue
        rec(G, v)

    # 帰りがけ順を逆順にする
    order.reverse()

    for v in order:
        print(f"{v} -> ", end="")
    print()


if __name__ == "__main__":
    main()
```

## code 13.9 木の DFS

木を根付き木として見て、各頂点の深さ depth と部分木サイズ subtree_size を求めます。

無向グラフとして辺を持つため、親 p に戻らないようにして DFS します。

```python
import sys


sys.setrecursionlimit(10 ** 7)

depth: list[int] = []
subtree_size: list[int] = []


def dfs(G: list[list[int]], v: int, p: int = -1, d: int = 0) -> None:
    # 頂点 v の深さを記録する
    depth[v] = d

    # 子へ進む。親 p へは戻らない
    for c in G[v]:
        if c == p:
            continue
        dfs(G, c, v, d + 1)

    # 子の部分木サイズが決まったあとで、自分の部分木サイズを計算する
    subtree_size[v] = 1
    for c in G[v]:
        if c == p:
            continue
        subtree_size[v] += subtree_size[c]


def main() -> None:
    global depth, subtree_size

    N = int(input())

    G = [[] for _ in range(N)]
    for _ in range(N - 1):
        a, b = map(int, input().split())
        G[a].append(b)
        G[b].append(a)

    root = 0
    depth = [0] * N
    subtree_size = [0] * N
    dfs(G, root)

    for v in range(N):
        print(f"{v}: depth = {depth[v]}, subtree_size = {subtree_size[v]}")


if __name__ == "__main__":
    main()
```
