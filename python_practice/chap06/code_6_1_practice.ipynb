## 1. アルゴリズムとコードの詳しい解説

```python

def binary_search(key: int) -> int:
    left = 0
    right = len(A) - 1
    while right >= left:
        mid = left + (right - left) // 2
        if A[mid] == key:
            return mid

        elif A[mid] < key:
```

### 目的

ソート済み配列 `A` から `key` を探し、見つかれば添字、なければ `-1` を返します。

### 変数・関数の意味

- `left`, `right`: 探索範囲の左右端です。
- `mid`: 中央の添字です。
- `key`: 探したい値です。
- `A`: ソート済み配列です。

### 処理の流れ

1. 中央を見て、目的値より大きければ左半分、小さければ右半分に探索範囲を狭めます。

### 計算量
