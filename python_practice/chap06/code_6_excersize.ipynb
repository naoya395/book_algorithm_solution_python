## 6.1　順位変換

### 問題文

長さ `N` の整数列 `A = (A_0, A_1, ..., A_{N-1})` が与えられます。

すべての要素の値は互いに異なります。

各 `i` について、`A_i` が数列全体の中で **何番目に小さい値か** を求めてください。

ただし、最も小さい値を `0` 番目、次に小さい値を `1` 番目、……と数えます。

### 制約

`1 <= N <= 2 * 10^5`

`0 <= A_i <= 10^9`

`A_i` はすべて相異なる

入力はすべて整数

### 要求計算量

`O(N log N)` 以内

### 要求空間計算量

`O(N)` 以内

### 入力

```text
N
A_0 A_1 ... A_{N-1}
```

### 出力

`N` 行出力してください。

`i` 行目には、`A_i` が小さい順で何番目かを出力してください。

### 入力例

```text
5
12 43 7 15 9
```

### 出力例

```text
2
4
0
3
1
```

### 説明

数列を小さい順に並べると、

```text
7, 9, 12, 15, 43
```

です。

したがって、`12` は `2` 番目、`43` は `4` 番目、`7` は `0` 番目です。


## 6.1　順位変換

考え方はシンプルです。

元の配列 `A` をコピーしてソートします。すると、ソート済み配列の中で `A[i]` が現れる位置が、その値の順位になります。

Python では、`bisect_left` を使えば「ソート済み配列の中でどこに入るか」を二分探索で求められます。全体計算量は、ソートに `O(N log N)`、各要素の二分探索に合計 `O(N log N)` なので、合計 `O(N log N)` です。

```python
from bisect import bisect_left

N = int(input())
A = list(map(int, input().split()))

B = sorted(A)

ans = []
for x in A:
    rank = bisect_left(B, x)
    ans.append(rank)

print(*ans, sep="\n")
```

ただし、今回の問題では値がすべて相異なるのでこれで十分です。もし重複がある一般的な座標圧縮なら、

```python
B = sorted(set(A))
```

とします。

## 6.2　3 つの屋台

条件は、

```text
A_i < B_j < C_k
```

です。

`B_j` を真ん中として固定して考えます。

ある値 `b = B_j` に対して、

`A_i < b` を満たす `A_i` の個数を `left`

`b < C_k` を満たす `C_k` の個数を `right`

とすると、この `b` を真ん中に使う組の個数は、

```text
left * right
```

です。

これをすべての `B_j` について足せば答えです。

`A`, `B`, `C` をソートしておけば、`left` は `bisect_left(A, b)` で求まります。`right` は、`C` の中で `b` より大きい要素数なので、`N - bisect_right(C, b)` です。

```python
from bisect import bisect_left, bisect_right

N = int(input())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
C = list(map(int, input().split()))

A.sort()
B.sort()
C.sort()

ans = 0

for b in B:
    left = bisect_left(A, b)
    right = N - bisect_right(C, b)
    ans += left * right

print(ans)
```

計算量は、ソートが `O(N log N)`、各 `B_j` に対して二分探索を 2 回するので `O(N log N)`、合計で `O(N log N)` です。

## 6.3　4 本のダーツ

4 個の値を直接全探索すると、組み合わせは `N^4` 個になってしまい、現実的ではありません。

そこで、4 個を

```text
前半2個 + 後半2個
```

に分けます。

まず、2 個選んだ和をすべて列挙します。

```text
P = [A_i + A_j for all i, j]
```

この配列 `P` の長さは `N^2` です。

あとは、

```text
P[x] + P[y] <= M
```

となる最大値を探せばよいです。

`P` をソートしておけば、ある `p` に対して、

```text
p + q <= M
```

を満たす最大の `q` を二分探索で探せます。

```python
from bisect import bisect_right

N, M = map(int, input().split())
A = list(map(int, input().split()))

pair_sums = []

for x in A:
    for y in A:
        pair_sums.append(x + y)

pair_sums.sort()

ans = 0

for p in pair_sums:
    limit = M - p
    idx = bisect_right(pair_sums, limit) - 1

    if idx >= 0:
        ans = max(ans, p + pair_sums[idx])

print(ans)
```

計算量は、2 個の和を作るのに `O(N^2)`、ソートに `O(N^2 log N)`、各 `p` に対して二分探索するので `O(N^2 log N)` です。合計で `O(N^2 log N)` です。

注意点として、Python では `N^2` 個の整数をリストで持つため、`N` が大きいとメモリが重くなります。たとえば `N=2000` なら `4,000,000` 個の整数を持つので、実行環境によってはギリギリになる可能性があります。アルゴリズムの学習としてはこの形が一番わかりやすいです。

## 6.4　離れた小屋

これは「最大値を直接求める」のではなく、

```text
距離を x 以上にして M 個選べるか？
```

という判定問題に変えます。

たとえば `x = 10` と決めたとき、小屋を左から順番に見ていき、最後に選んだ小屋から `10` 以上離れていたら選ぶ、という貪欲法をします。

この方法で `M` 個以上選べたら、距離 `x` は達成可能です。

達成可能ならもっと大きい距離を試したい。

達成不可能なら距離を小さくする必要がある。

つまり、答えに対して二分探索できます。

```python
N, M = map(int, input().split())
X = list(map(int, input().split()))

X.sort()

def can_place(dist):
    count = 1
    last = X[0]

    for i in range(1, N):
        if X[i] - last >= dist:
            count += 1
            last = X[i]

    return count >= M

left = 0
right = X[-1] - X[0] + 1

while right - left > 1:
    mid = (left + right) // 2

    if can_place(mid):
        left = mid
    else:
        right = mid

print(left)
```

`can_place(dist)` は `O(N)` です。二分探索は座標の最大差を `A` とすると `O(log A)` 回なので、全体計算量は `O(N log A)` です。

この問題で重要なのは、「最小距離の最大化」という言い方に惑わされず、

```text
その距離を達成できるか？
```

に変換することです。

## 6.5　掛け算表の K 番目

`N^2` 個の積をすべて作ってソートすると、`N` が大きい場合は間に合いません。

そこで、値 `x` を決めて、

```text
A_i * B_j <= x
```

を満たす組の個数を数えます。

この個数が `K` 個以上なら、`K` 番目に小さい値は `x` 以下です。

この個数が `K` 個未満なら、`K` 番目に小さい値は `x` より大きいです。

つまり、答えの値そのものを二分探索できます。

`B` をソートしておくと、ある `a = A_i` に対して、

```text
a * B_j <= x
```

は、

```text
B_j <= x // a
```

と同じです。

したがって、`bisect_right(B, x // a)` で個数を数えられます。

```python
from bisect import bisect_right

N, K = map(int, input().split())
A = list(map(int, input().split()))
B = list(map(int, input().split()))

A.sort()
B.sort()

def count_leq(x):
    count = 0

    for a in A:
        count += bisect_right(B, x // a)

    return count

left = 0
right = A[-1] * B[-1] + 1

while right - left > 1:
    mid = (left + right) // 2

    if count_leq(mid) >= K:
        right = mid
    else:
        left = mid

print(right)
```

この二分探索では、

`left` は「条件を満たさない値」

`right` は「条件を満たす値」

として管理しています。

最終的に、`count_leq(x) >= K` を初めて満たす最小の `x` が答えです。

計算量は、`B` のソートで `O(N log N)`、判定 1 回につき `O(N log N)`、値の範囲を `C` とすると二分探索が `O(log C)` 回なので、全体で `O(N log N log C)` です。

## 6.6　揺れるボール

方程式は、

```text
A t + B sin(Cπt) = 100
```

です。

左辺を

```python
f(t) = A * t + B * sin(C * pi * t)
```

とします。

すると、

```text
f(t) - 100 = 0
```

となる `t` を探す問題です。

実数の答えを求めるので、整数の二分探索ではなく、実数の二分探索をします。

まず、`t = 0` のとき、

```text
f(0) = 0
```

なので `f(0) - 100 < 0` です。

一方で、

```text
t = (100 + B) / A
```

とすると、`sin` の値が最悪で `-1` でも、

```text
A t + B sin(Cπt) >= A t - B = 100
```

となります。

したがって、この範囲の中に少なくとも 1 つ解があります。

```python
import math

A, B, C = map(int, input().split())

def f(t):
    return A * t + B * math.sin(C * math.pi * t)

left = 0.0
right = (100.0 + B) / A

for _ in range(100):
    mid = (left + right) / 2

    if f(mid) >= 100:
        right = mid
    else:
        left = mid

print(right)
```

実数二分探索では、`while right - left > 1e-7` のようにしてもよいですが、競プロでは固定回数回す実装もよく使います。

`100` 回回せば区間幅は十分小さくなるので、`10^-6` 程度の誤差条件にはかなり余裕があります。
