```python
                ok = False

        if ok:
            right = mid
        else:
            left = mid

    print(right)

```

### 目的

スコア上限 `mid` 以内に全風船を割れるか判定し、可能な最小スコアを二分探索で求めます。

### 変数・関数の意味

- `h[i]`: 風船 `i` の初期高度です。
- `s[i]`: 風船 `i` の上昇速度です。
- `t[i]`: スコア `mid` 以下で割るための締切時刻です。
- `ok`: `mid` が実現可能かどうかです。

### 処理の流れ

1. 各風船の締切 `t[i]` を計算し、早い順に並べます。`i` 番目に割る風船の締切が `i` 未満なら不可能です。

### 計算量

時間計算量は **O(n log n log M)** です。


## 問題文

射撃大会に参加したあなたは、空へ向かって上昇していく風船をすべて撃ち落とすことになりました。

風船は全部で $N$ 個あります。風船 $i$ は、時刻 $0$ の時点で高さ $H_i$ にあり、そこから $1$ 秒ごとに高さが $S_i$ ずつ増加します。つまり、時刻 $t$ 秒における風船 $i$ の高さは

$H_i + S_i t$

です。

あなたは時刻 $0$ にちょうど 1 個の風船を撃つことができ、その後も時刻 $1, 2, \ldots$ にそれぞれ 1 個ずつ風船を撃つことができます。風船を撃つ順番は自由に決められます。

風船を撃ったとき、その時点での風船の高さがペナルティになります。すべての風船を撃ち落としたとき、各風船を撃った際のペナルティの最大値を最終ペナルティとします。

最終ペナルティとして達成できる最小値を求めてください。

## 入力

入力は以下の形式で標準入力から与えられます。

```text
N
H_1 S_1
H_2 S_2
\vdots
H_N S_N
```

## 出力

最終ペナルティとして達成できる最小値を 1 行に出力してください。

## 制約

- $1 \leq N \leq 10^5$
- $1 \leq H_i \leq 10^9$
- $1 \leq S_i \leq 10^9$
- 入力はすべて整数

## 入力例 1

```text
4
5 6
12 4
14 7
21 2
```

## 出力例 1

```text
23
```

## 説明

例えば、次の順に風船を撃つことを考えます。

- 時刻 $0$ に風船 1 を撃つ: 高さ $5$
- 時刻 $1$ に風船 2 を撃つ: 高さ $16$
- 時刻 $2$ に風船 4 を撃つ: 高さ $25$
- 時刻 $3$ に風船 3 を撃つ: 高さ $35$

この場合の最終ペナルティは $35$ です。

しかし、よりよい順番を選ぶことで、最終ペナルティを $23$ 以下にできます。最終ペナルティを $22$ 以下にすることはできないため、答えは $23$ です。
