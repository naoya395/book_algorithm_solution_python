## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n = int(input())

    s = [0] * n
    for i in range(n):
        h[i], s[i] = map(int, input().split())

    m = 0
    for i in range(n):
        m = max(m, h[i] + s[i] * n)

    left = 0
    right = m
    while right - left > 1:
        mid = (left + right) // 2

        ok = True
        t = [0] * n
        for i in range(n):
            if mid < h[i]:
                ok = False
            else:
                t[i] = (mid - h[i]) // s[i]

        t.sort()
        for i in range(n):
            if t[i] < i:
                ok = False

        if ok:
            right = mid
        else:
            left = mid

    print(right)


if __name__ == "__main__":
    main()
```

### 目的

スコア上限 `mid` 以内に全風船を割れるか判定し、可能な最小スコアを二分探索で求めます。

### 変数・関数の意味

- `h[i]`: 風船 `i` の初期高度です。
- `s[i]`: 風船 `i` の上昇速度です。
- `t[i]`: スコア `mid` 以下で割るための締切時刻です。
- `ok`: `mid` が実現可能かどうかです。

### 処理の流れ

1. 各風船の締切 `t[i]` を計算し、早い順に並べます。`i` 番目に割る風船の締切が `i` 未満なら不可能です。

### 計算量

時間計算量は **O(n log n log M)** です。
