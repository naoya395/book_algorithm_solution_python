質問への yes/no を使って、年齢が入る範囲 `[left, right)` を狭めます。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    print("Start Game!")

    left = 20
    right = 36

    while right - left > 1:
        mid = left + (right - left) // 2

        print(f"Is the age less than {mid} ? (yes / no)")
        ans = input()

        if ans == "yes":
            right = mid
        else:
            left = mid

    print(f"The age is {left}!")


if __name__ == "__main__":
    main()
```

### 目的

質問への yes/no を使って、年齢が入る範囲 `[left, right)` を狭めます。

### 変数・関数の意味

- `left`: 年齢がこれより大きいことが分かっている境界です。
- `right`: 年齢がこれより小さいことが分かっている境界です。
- `mid`: 質問に使う中央の値です。
- `ans`: ユーザーの回答です。

### 処理の流れ

1. `mid` より小さいなら `right = mid`、そうでなければ `left = mid` として範囲を半分にします。

### 計算量

時間計算量は **O(log 範囲幅)** です。
