## 1. アルゴリズムとコードの詳しい解説

```python
from bisect import bisect_left

INF = 20000000


def main() -> None:
    n, k = map(int, input().split())
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))

    min_value = INF

    b.sort()
    b.append(INF)

    for i in range(n):
        index = bisect_left(b, k - a[i])
        val = b[index]

        if a[i] + val < min_value:
            min_value = a[i] + val

    print(min_value)


if __name__ == "__main__":
    main()
```

### 目的

`b` をソートし、各 `a[i]` に対して `k - a[i]` 以上の最小の `b` を `bisect_left` で探します。

### 変数・関数の意味

- `bisect_left`: ソート済み配列で挿入位置を返します。
- `k`: 和の下限です。
- `min_value`: 条件を満たす最小和です。
- `val`: 今回選んだ `b` の値です。

### 処理の流れ

1. `b` をソートして番兵 `INF` を追加し、各 `a[i]` に対して必要な `b` を二分探索で見つけます。

### 計算量

時間計算量は **O(n log n)** です。
