`b` をソートし、各 `a[i]` に対して `k - a[i]` 以上の最小の `b` を `bisect_left` で探します。


## 1. アルゴリズムとコードの詳しい解説

```python
from bisect import bisect_left

INF = 20000000


def main() -> None:
    n, k = map(int, input().split())
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))

    min_value = INF

    b.sort()
    b.append(INF)

    for i in range(n):
        index = bisect_left(b, k - a[i])
        val = b[index]

        if a[i] + val < min_value:
            min_value = a[i] + val

    print(min_value)


if __name__ == "__main__":
    main()
```

### 目的

`b` をソートし、各 `a[i]` に対して `k - a[i]` 以上の最小の `b` を `bisect_left` で探します。

### 変数・関数の意味

- `bisect_left`: ソート済み配列で挿入位置を返します。
- `k`: 和の下限です。
- `min_value`: 条件を満たす最小和です。
- `val`: 今回選んだ `b` の値です。

### 処理の流れ

1. `b` をソートして番兵 `INF` を追加し、各 `a[i]` に対して必要な `b` を二分探索で見つけます。

### 計算量

時間計算量は **O(n log n)** です。


## 問題文：目標以上になる最小の合計

あなたはイベントで使うために、2種類の部品を1つずつ選んで装置を作ろうとしています。

部品Aは $N$ 個あり、それぞれの性能値は

$$
a_1, a_2, \dots, a_N
$$

です。

部品Bも $N$ 個あり、それぞれの性能値は

$$
b_1, b_2, \dots, b_N
$$

です。

部品Aから1つ、部品Bから1つを選んだとき、装置全体の性能値はその和になります。

性能値の合計が $K$ 以上になるように部品を選びたいです。
そのような選び方のうち、合計値が最小となるものを求めてください。

つまり、次を満たす最小値を出力してください。

$$
\min(a_i + b_j) \quad \text{subject to} \quad a_i + b_j \ge K
$$




## 使用してよい機能

この問題では、Python の次の機能を使ってよいものとします。

```python
from bisect import bisect_left
```

また、リストの `sort()` メソッドも使ってよいです。

```python
B.sort()
```

## `sort()` の使い方

`sort()` はリストを昇順に並べ替えるメソッドです。

```python
B = [9, 4, 13, 5]
B.sort()
print(B)
```

出力は次のようになります。

```text
[4, 5, 9, 13]
```

二分探索を使うためには、探索対象の配列がソートされている必要があります。

## `bisect_left` の使い方

`bisect_left(B, x)` は、ソート済みのリスト `B` に対して、`x` 以上の値が初めて現れる位置を返します。

例えば、

```python
from bisect import bisect_left

B = [4, 5, 9, 13, 21]

print(bisect_left(B, 10))
print(bisect_left(B, 13))
print(bisect_left(B, 100))
```

出力は次のようになります。

```text
3
3
5
```

`10` 以上の最初の値は `13` なので、インデックス `3` が返ります。
`13` はそのまま存在するので、インデックス `3` が返ります。
`100` 以上の値は存在しないので、リストの長さである `5` が返ります。

この問題では、各 `A_i` に対して、

```python
B_j >= K - A_i
```

となる最小の `B_j` を探せばよいです。

そのため、

```python
index = bisect_left(B, K - A_i)
```

のようにして候補を見つけられます。

```text
N K
a_1 a_2 ... a_N

```

---

## 出力形式

条件を満たす $a_i + b_j$ の最小値を1行で出力してください。

---

## 制約

このコードがそのまま安全に動くためには、例えば次のような制約が必要です。

```text
1 <= N <= 200000
0 <= a_i, b_i <= 10000000
0 <= K <= 20000000
条件を満たす組 (i, j) が少なくとも1つ存在する
入力はすべて整数
```

特に重要なのは、コード内で `INF = 20000000` を番兵として使っている点です。

そのため、$K$ や $a_i$, $b_i$ がこれより大きくなる可能性がある問題では、このコードは正しく動かない場合があります。
