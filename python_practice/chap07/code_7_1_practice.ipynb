## 1. アルゴリズムとコードの詳しい解説

```python
VALUE = [500, 100, 50, 10, 5, 1]


def main() -> None:
    x = int(input())
    a = list(map(int, input().split()))

    result = 0
    for i in range(6):
        add = x // VALUE[i]

        if add > a[i]:
            add = a[i]

        x -= VALUE[i] * add
        result += add

    print(result)


if __name__ == "__main__":
    main()
```

### 目的

大きい額面から順に、使えるだけ使って支払う枚数を数えます。

### 変数・関数の意味

- `VALUE`: 硬貨の額面を大きい順に並べた配列です。
- `x`: まだ支払う必要がある金額です。
- `a[i]`: 額面 `VALUE[i]` の硬貨の枚数です。
- `add`: 今回使う硬貨枚数です。
- `result`: 使った硬貨の総枚数です。

### 処理の流れ

1. 各額面について `x // VALUE[i]` 枚まで使えますが、手持ち枚数 `a[i]` を超えないようにします。使った分だけ `x` を減らし、`result` を増やします。

### 計算量

時間計算量は **O(1)**（額面数が固定のため） です。
