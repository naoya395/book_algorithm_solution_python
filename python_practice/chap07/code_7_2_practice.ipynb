## 1. アルゴリズムとコードの詳しい解説


### 目的

終了時刻が早い順に区間を選ぶことで、重ならずに選べる区間数を最大化します。

### 変数・関数の意味

- `inter`: `(start, end)` の区間リストです。
- `current_end_time`: 最後に選んだ区間の終了時刻です。
- `res`: 選んだ区間数です。

### 処理の流れ

1. 区間を終了時刻でソートし、開始時刻が現在の終了時刻以上なら採用します。

### 計算量

時間計算量は **O(n log n)** です。


## 問題文

あなたは、今日こなす仕事の候補をいくつか持っています。

仕事は全部で $N$ 個あり、$i$ 番目の仕事は時刻 $s_i$ に始まり、時刻 $t_i$ に終わります。
同時に複数の仕事を行うことはできないため、選んだ仕事同士の時間が重なってはいけません。

ただし、ある仕事が時刻 $t$ に終わり、別の仕事が同じ時刻 $t$ に始まる場合、この 2 つの仕事は両方選ぶことができます。

できるだけ多くの仕事を選ぶとき、最大で何個の仕事をこなすことができるか求めてください。

## 入力

入力は次の形式で標準入力から与えられます。

```text
N
s_0 t_0
s_1 t_1
...
s_{N-1} t_{N-1}
```

## 出力

重ならないように選べる仕事の最大個数を 1 行に出力してください。

## 制約

- $1 \leq N \leq 2 \times 10^5$
- $0 \leq s_i < t_i \leq 10^9$
- 入力はすべて整数である

## 入力例 1

```text
5
1 3
2 5
4 7
6 9
8 10
```

## 出力例 1

```text
3
```

## 説明 1

例えば、次の 3 個の仕事を選ぶことができます。

- 時刻 $1$ から $3$ までの仕事
- 時刻 $4$ から $7$ までの仕事
- 時刻 $8$ から $10$ までの仕事

4 個以上の仕事を重ならないように選ぶことはできないため、答えは `3` です。

## 入力例 2

```text
6
0 2
2 4
1 3
4 6
5 7
6 8
```

## 出力例 2

```text
4
```

## 説明 2

時刻がちょうど接している仕事は重ならないものとして扱います。

例えば、次の 4 個の仕事を選べます。

- 時刻 $0$ から $2$ までの仕事
- 時刻 $2$ から $4$ までの仕事
- 時刻 $4$ から $6$ までの仕事
- 時刻 $6$ から $8$ までの仕事
