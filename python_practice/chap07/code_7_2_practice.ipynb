終了時刻が早い順に区間を選ぶことで、重ならずに選べる区間数を最大化します。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n = int(input())
    inter = []
    for _ in range(n):
        start, end = map(int, input().split())
        inter.append((start, end))

    inter.sort(key=lambda interval: interval[1])

    res = 0
    current_end_time = 0
    for i in range(n):
        if inter[i][0] < current_end_time:
            continue

        res += 1
        current_end_time = inter[i][1]

    print(res)


if __name__ == "__main__":
    main()
```

### 目的

終了時刻が早い順に区間を選ぶことで、重ならずに選べる区間数を最大化します。

### 変数・関数の意味

- `inter`: `(start, end)` の区間リストです。
- `current_end_time`: 最後に選んだ区間の終了時刻です。
- `res`: 選んだ区間数です。

### 処理の流れ

1. 区間を終了時刻でソートし、開始時刻が現在の終了時刻以上なら採用します。

### 計算量

時間計算量は **O(n log n)** です。
