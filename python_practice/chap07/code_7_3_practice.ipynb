後ろの条件から順に、`a[i] + total` が `b[i]` の倍数になるよう追加量を決めます。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n = int(input())
    a = [0] * n
    b = [0] * n
    for i in range(n):
        a[i], b[i] = map(int, input().split())

    total = 0
    for i in range(n - 1, -1, -1):
        a[i] += total
        amari = a[i] % b[i]
        d = 0
        if amari != 0:
            d = b[i] - amari
        total += d

    print(total)


if __name__ == "__main__":
    main()
```

### 目的

後ろの条件から順に、`a[i] + total` が `b[i]` の倍数になるよう追加量を決めます。

### 変数・関数の意味

- `a[i]`, `b[i]`: 各条件の値です。
- `total`: これまで後ろ側で追加した総量です。
- `amari`: `a[i] + total` を `b[i]` で割った余りです。
- `d`: 今回追加する量です。

### 処理の流れ

1. 後ろから前へ見て、余りが 0 でなければ次の倍数までの差 `b[i] - amari` を追加します。

### 計算量

時間計算量は **O(n)** です。



## 問題文

長さ $N$ の整数列 $A=(A_0,A_1,\ldots,A_{N-1})$ と、長さ $N$ の正整数列 $B=(B_0,B_1,\ldots,B_{N-1})$ が与えられます。

また、$N$ 個のボタンがあります。$j$ 番目のボタンを 1 回押すと、$A_0,A_1,\ldots,A_j$ の値がそれぞれ $1$ ずつ増えます。

あなたはボタンを何回か押して、すべての $i$ について $A_i$ が $B_i$ の倍数になるようにしたいです。

この条件を満たすために必要なボタンを押す回数の最小値を求めてください。

## 入力

入力は次の形式で標準入力から与えられます。

```text
N
A_0 B_0
A_1 B_1
\vdots
A_{N-1} B_{N-1}
```

## 出力

すべての $i$ について $A_i$ が $B_i$ の倍数になるようにするために必要な、ボタンを押す回数の最小値を出力してください。

## 制約

- $1 \leq N \leq 10^5$
- $0 \leq A_i \leq 10^9$
- $1 \leq B_i \leq 10^9$
- 入力はすべて整数である。

## 入力例 1

```text
3
2 3
3 5
4 7
```

## 出力例 1

```text
7
```

## 説明 1

例えば、まず $2$ 番目のボタンを $3$ 回押すと、$A_2$ は $7$ になり、$B_2=7$ の倍数になります。

その後、$1$ 番目のボタンを $4$ 回押すと、$A_1$ は $10$ になり、$B_1=5$ の倍数になります。この時点で $A_0$ は $9$ であり、$B_0=3$ の倍数です。

したがって、合計 $7$ 回の操作で条件を満たすことができます。

## 入力例 2

```text
3
1 2
3 2
4 4
```

## 出力例 2

```text
1
```

## 説明 2

$1$ 番目のボタンを $1$ 回押すと、$A=(2,4,4)$ になります。

このとき、すべての $i$ について $A_i$ は $B_i$ の倍数です。
