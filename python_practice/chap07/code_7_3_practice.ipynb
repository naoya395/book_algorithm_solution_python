後ろの条件から順に、`a[i] + total` が `b[i]` の倍数になるよう追加量を決めます。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n = int(input())
    a = [0] * n
    b = [0] * n
    for i in range(n):
        a[i], b[i] = map(int, input().split())

    total = 0
    for i in range(n - 1, -1, -1):
        a[i] += total
        amari = a[i] % b[i]
        d = 0
        if amari != 0:
            d = b[i] - amari
        total += d

    print(total)


if __name__ == "__main__":
    main()
```

### 目的

後ろの条件から順に、`a[i] + total` が `b[i]` の倍数になるよう追加量を決めます。

### 変数・関数の意味

- `a[i]`, `b[i]`: 各条件の値です。
- `total`: これまで後ろ側で追加した総量です。
- `amari`: `a[i] + total` を `b[i]` で割った余りです。
- `d`: 今回追加する量です。

### 処理の流れ

1. 後ろから前へ見て、余りが 0 でなければ次の倍数までの差 `b[i] - amari` を追加します。

### 計算量

時間計算量は **O(n)** です。
