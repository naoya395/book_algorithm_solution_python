
# 7.1　大小ペア

この問題では、`A` から選んだ値 `a` と `B` から選んだ値 `b` について、`a < b` ならペアを作れます。各要素は 1 回しか使えません。最大で何個ペアを作れるかを求めます。

まず、`A` と `B` を両方とも小さい順にソートします。

その後、`B` を小さい順に見ていきます。今まだ使っていない `A` のうち、最も小さいものを `A[i]` とします。

現在見ている `B` の値を `b` とすると、次のように考えます。

`A[i] < b` なら、`A[i]` と `b` はペアにできます。このとき、`A[i]` は今使っていない `A` の中で最も小さい値なので、これを使ってしまうのがよいです。ペアを 1 個作り、`i` を 1 つ進めます。

一方、`A[i] >= b` なら、`A[i]` ですら `b` とペアにできません。`A` はソートされているので、`A[i]` より後ろの値はすべて `A[i]` 以上です。つまり、どの未使用の `A` もこの `b` とはペアにできません。したがって、この `b` は諦めて次の `B` を見ます。

たとえば、

```text
A = [1, 2, 4, 6, 8]
B = [2, 3, 5, 7, 9]
```

だとします。

小さい順に `B` を見ます。

`b = 2` のとき、最小の未使用 `A` は `1` です。`1 < 2` なのでペアにできます。

`b = 3` のとき、最小の未使用 `A` は `2` です。`2 < 3` なのでペアにできます。

`b = 5` のとき、最小の未使用 `A` は `4` です。`4 < 5` なのでペアにできます。

このように、今作れるなら作る、作れないならその `b` は捨てる、という方針です。

なぜこの貪欲法でよいかを考えます。

ポイントは、「小さい `B` から順に処理する」というところです。小さい `B` は条件が厳しいです。なぜなら `a < b` を満たすためには、`a` がかなり小さくないといけないからです。

逆に、大きい `B` は条件が緩いです。多くの `A` とペアにできます。

したがって、条件が厳しい小さい `B` から順に、使える `A` を割り当てていくのが自然です。

さらに、ある `b` に対してペアを作れるなら、未使用の `A` のうち最も小さいものを使ってよいです。

なぜなら、もしある最適な作り方で、現在の `b` に対して `A[i]` より大きい `A[k]` を使っていたとします。このとき、`A[i]` も `b` とペアにできます。そこで、現在の `b` には `A[i]` を使うように変更しても損しません。

もし `A[i]` が別のもっと大きい `B` とペアになっていたとしても、元々 `b` に使っていた `A[k]` は `A[i]` より大きいとはいえ、`A[k] < b` が成り立っています。そして、その別の `B` は現在の `b` 以上の値です。したがって `A[k]` もその別の `B` とペアにできます。つまり、ペアの組み替えが可能です。

このように、「今の `b` に対して、最小の使える `A` を使う」ような最適解が必ず存在します。よって、この操作を繰り返せば最大個数が得られます。

注意点として、`i == N` になったら、もう使える `A` は残っていません。その後に `A[i]` を見ようとすると Python ではエラーになります。実装では `i < N` を必ず確認します。

計算量は、ソートに `O(N log N)`、その後の走査に `O(N)` かかるので、全体で `O(N log N)` です。空間計算量は、入力配列を持つので `O(N)` です。

```python
import sys

def main():
    input = sys.stdin.readline

    N = int(input())
    A = list(map(int, input().split()))
    B = list(map(int, input().split()))

    A.sort()
    B.sort()

    i = 0
    ans = 0

    for b in B:
        if i < N and A[i] < b:
            ans += 1
            i += 1

    print(ans)

if __name__ == "__main__":
    main()
```

# 7.2　赤と青の支配ペア

この問題では、赤い点 `(rx, ry)` と青い点 `(bx, by)` について、

```text
rx < bx かつ ry < by
```

が成り立つとき、ペアを作れます。各点は 1 回しか使えません。最大で何個ペアを作れるかを求めます。

一次元の 7.1 と違って、今回は `x` 座標と `y` 座標の両方を考える必要があります。

基本方針は、青い点を `x` 座標の小さい順に見ていくことです。

ある青い点 `(bx, by)` を見ているとします。この青い点とペアにできる赤い点は、

```text
rx < bx
ry < by
```

を満たす赤い点です。

青い点を `bx` の小さい順に処理していくと、ある時点で `rx < bx` を満たす赤い点だけを候補に追加していくことができます。

つまり、処理の流れはこうです。

赤い点を `x` 座標の小さい順にソートします。

青い点も `x` 座標の小さい順にソートします。

青い点を左から右へ見ていきます。

現在の青い点の `bx` より小さい `x` 座標を持つ赤い点を、候補集合に追加します。

その候補集合の中から、`y < by` を満たす赤い点を 1 つ選べば、現在の青い点とペアにできます。

ここで大事なのは、「どの赤い点を選ぶべきか」です。

候補の赤い点のうち、`y < by` を満たすものが複数あるなら、その中で `y` が最大のものを選びます。

つまり、現在の青い点に入る赤い点のうち、できるだけ上にある赤い点を使います。

なぜなら、`y` が小さい赤い点は、今後出てくる `by` が小さい青い点にも使いやすいからです。逆に、`y` が大きい赤い点は、使える相手が限られます。したがって、今使えるギリギリ大きい `y` の赤い点を使い、小さい `y` の赤い点を将来に残すのが得です。

たとえば、現在の青い点の `by` が `10` で、候補の赤い点の `y` が

```text
2, 4, 7, 9
```

だとします。

この場合、すべて `10` より小さいので現在の青い点に使えます。しかし、ここで `2` を使ってしまうのはもったいないです。将来 `by = 3` のような青い点が出てきた場合、`2` は使えますが、`4, 7, 9` は使えません。

そのため、現在は `9` を使うのがよい、という考え方です。

この貪欲法の正しさをもう少し丁寧に見ます。

青い点を `x` 座標の小さい順に処理しているので、現在の青い点より後に出てくる青い点は、現在の青い点以上の `x` 座標を持ちます。

したがって、現在候補に入っている赤い点、つまり `rx < bx` を満たしている赤い点は、今後の青い点に対しても `x` 座標の条件を満たしやすいです。問題は主に `y` 座標になります。

現在の青い点に対して、`y < by` を満たす赤い点があるなら、現在の青い点は誰かとペアにしてよいです。仮に最適解で現在の青い点がペアになっていなかったとしても、候補の赤い点が余っていればペアを増やせてしまいます。候補の赤い点が将来の青い点とペアになっていたとしても、それを現在の青い点に移すことで、ペア数を減らさない形にできます。

さらに、現在の青い点に使う赤い点として、`y < by` を満たす中で最大の `y` のものを選んでも大丈夫です。

もし最適解が、現在の青い点にもっと小さい `y` の赤い点を使っていたとします。そして、今回の貪欲法が選ぼうとしている大きい `y` の赤い点が、別の未来の青い点とペアになっていたとします。

その未来の青い点は、現在の青い点より右側にあります。また、貪欲法が選ぶ赤い点の `y` は、その未来の青い点の `by` より小さいからこそ、未来の青い点とペアになっていたはずです。

一方、元々現在の青い点に使われていた赤い点は、それより `y` が小さいです。したがって、その未来の青い点にも使えます。

つまり、

```text
現在の青い点 ← 大きい y の赤い点
未来の青い点 ← 小さい y の赤い点
```

というように交換しても、ペア数は減りません。

これにより、現在の青い点には「使える中で最大の `y` の赤い点」を使う貪欲法が正しいと分かります。

実装では、「候補集合の中から `by` 未満で最大の `y` を持つ赤い点を取り出す」必要があります。

C++ なら `multiset` で簡単にできますが、Python の標準ライブラリには平衡二分探索木がありません。`list` に入れて `bisect` で探し、削除を `pop` すると、削除が `O(N)` になってしまいます。

そこで、Python では Fenwick Tree、つまり BIT を使うのが安全です。

やりたいことは、赤い点の `y` 座標を管理することです。

まず、赤い点の `y` 座標を座標圧縮します。たとえば赤い点の `y` が

```text
100, 20, 100, 50
```

なら、種類だけ取り出してソートし、

```text
20, 50, 100
```

とします。

それぞれを番号に変換します。

```text
20  -> 0
50  -> 1
100 -> 2
```

BIT には、「現在候補集合にある赤い点の `y` が何個あるか」を記録します。

現在の青い点の `by` に対して、`by` 未満の `y` の種類数を `k` とします。これは `bisect_left(ys, by)` で求められます。

BIT で `[0, k)` の範囲に候補が何個あるかを調べます。これが 0 なら、現在の青い点とペアにできる赤い点はありません。

1 個以上あるなら、その範囲にある候補のうち最も大きい `y` を選びたいです。これは、「`by` 未満の候補数」を `cnt` としたとき、候補全体の中で `cnt` 番目に小さい `y` を探せばよいです。BIT の `lower_bound` を使うと、これを `O(log N)` で探せます。

計算量は、ソートが `O(N log N)`、各点の追加と削除がそれぞれ `O(log N)` なので、全体で `O(N log N)` です。空間計算量は、点の配列と BIT を持つので `O(N)` です。

```python
import sys
from bisect import bisect_left

class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.data = [0] * (n + 1)

    def add(self, i, x):
        # i は 0-indexed
        i += 1
        while i <= self.n:
            self.data[i] += x
            i += i & -i

    def sum(self, r):
        # [0, r) の合計
        s = 0
        while r > 0:
            s += self.data[r]
            r -= r & -r
        return s

    def lower_bound(self, w):
        """
        累積和が初めて w 以上になる位置を返す。
        返り値は 0-indexed。
        w >= 1 を仮定する。
        """
        x = 0
        k = 1 << (self.n.bit_length() - 1)

        while k > 0:
            if x + k <= self.n and self.data[x + k] < w:
                w -= self.data[x + k]
                x += k
            k >>= 1

        return x

def main():
    input = sys.stdin.readline

    N = int(input())

    red = []
    for _ in range(N):
        x, y = map(int, input().split())
        red.append((x, y))

    blue = []
    for _ in range(N):
        x, y = map(int, input().split())
        blue.append((x, y))

    red.sort()
    blue.sort()

    ys = sorted(set(y for _, y in red))
    bit = FenwickTree(len(ys))

    ans = 0
    p = 0

    for bx, by in blue:
        # x 座標が bx より小さい赤い点を候補に追加する
        while p < N and red[p][0] < bx:
            ry = red[p][1]
            idx = bisect_left(ys, ry)
            bit.add(idx, 1)
            p += 1

        # by より小さい y 座標を持つ候補が何個あるか
        limit = bisect_left(ys, by)
        cnt = bit.sum(limit)

        if cnt > 0:
            # by 未満の候補のうち、最も y が大きいものを選ぶ
            idx = bit.lower_bound(cnt)
            bit.add(idx, -1)
            ans += 1

    print(ans)

if __name__ == "__main__":
    main()
```

この実装では、同じ `y` 座標を持つ赤い点が複数あっても大丈夫です。BIT には個数を入れているので、同じ座標が複数あれば、その個数分だけ管理できます。

また、条件は `rx < bx` と `ry < by` であり、どちらも「未満」です。実装では、赤い点を候補に追加するときに

```python
red[p][0] < bx
```

を使っています。また、`y` 座標については

```python
bisect_left(ys, by)
```

を使うことで、`by` と等しい `y` は含めず、`by` 未満だけを候補にしています。

# 7.3　締め切り付きタスク

この問題では、`N` 個の仕事があります。仕事 `i` は `d_i` 時間かかり、締め切りが `t_i` です。

時刻 `0` から始めて、仕事を 1 つずつ実行します。同時に複数の仕事はできません。すべての仕事を、それぞれの締め切りまでに終えられるかを判定します。

結論から言うと、締め切り `t_i` が早い順に仕事を並べて、その順番で実行すればよいです。

そして、順番に処理していき、現在時刻がその仕事の締め切りを超えたら `No`、最後まで超えなければ `Yes` です。

たとえば、次の 3 つの仕事があるとします。

```text
仕事A: かかる時間 2, 締め切り 8
仕事B: かかる時間 4, 締め切り 9
仕事C: かかる時間 3, 締め切り 5
```

締め切りが早い順に並べると、

```text
仕事C: 3, 5
仕事A: 2, 8
仕事B: 4, 9
```

です。

時刻 0 から始めます。

仕事C をすると、時刻は `3` になります。締め切り `5` 以下なので大丈夫です。

仕事A をすると、時刻は `5` になります。締め切り `8` 以下なので大丈夫です。

仕事B をすると、時刻は `9` になります。締め切り `9` 以下なので大丈夫です。

したがって `Yes` です。

では、なぜ締め切りが早い順でよいのでしょうか。

これは「隣り合う 2 つの仕事の交換」で説明できます。

あるスケジュールの中で、隣り合う 2 つの仕事 `A` と `B` がこの順で並んでいるとします。

```text
A → B
```

しかし、締め切りは `B` の方が早いとします。

```text
t_A > t_B
```

つまり、本来は `B` を先にやった方が自然です。

このとき、`A → B` の順で両方の仕事が締め切りに間に合っているなら、`B → A` に入れ替えても締め切りに間に合います。

仕事 `A` の所要時間を `d_A`、締め切りを `t_A` とします。仕事 `B` の所要時間を `d_B`、締め切りを `t_B` とします。また、この 2 つの仕事を始める前の時刻を `S` とします。

元の順番 `A → B` では、完了時刻は次のようになります。

```text
A の完了時刻: S + d_A
B の完了時刻: S + d_A + d_B
```

この順番で両方が間に合っているなら、

```text
S + d_A <= t_A
S + d_A + d_B <= t_B
```

が成り立っています。

次に、順番を `B → A` に変えます。

```text
B の完了時刻: S + d_B
A の完了時刻: S + d_B + d_A
```

まず `B` について考えると、

```text
S + d_B <= S + d_A + d_B
```

です。そして元々、

```text
S + d_A + d_B <= t_B
```

だったので、

```text
S + d_B <= t_B
```

が成り立ちます。よって `B` は間に合います。

次に `A` について考えます。

`A` の完了時刻は

```text
S + d_B + d_A
```

です。これは元の順番での `B` の完了時刻と同じです。元々、

```text
S + d_A + d_B <= t_B
```

でした。そして、`t_B < t_A` なので、

```text
S + d_A + d_B <= t_A
```

も成り立ちます。よって `A` も間に合います。

つまり、締め切りが遅い仕事が前にあり、締め切りが早い仕事が後ろにある場合、その 2 つを入れ替えても悪くなりません。

この交換を何度も繰り返すと、最終的に締め切りが早い順に並んだスケジュールになります。そして、その途中で「間に合うスケジュール」が「間に合わないスケジュール」に変わることはありません。

したがって、もしどこかに全仕事を間に合わせる順番が存在するなら、締め切りが早い順に並べた順番でも間に合います。

逆に、締め切りが早い順にやっても間に合わないなら、どんな順番でも間に合いません。

よって、締め切りが早い順に並べて確認すれば十分です。

実装はかなりシンプルです。

仕事を `(t_i, d_i)` の形で持ち、締め切り `t_i` でソートします。

現在時刻 `now` を `0` にします。

各仕事について、

```python
now += d
```

として完了時刻を求めます。

もし

```python
now > t
```

なら、その仕事は締め切りを過ぎてから終わるので `No` です。

すべての仕事を確認して一度も締め切りを超えなければ `Yes` です。

ここで、完了時刻が締め切りとちょうど同じ場合は間に合っています。したがって、判定は `now > t` のときだけ `No` です。`now == t` は大丈夫です。

計算量は、ソートが `O(N log N)`、確認が `O(N)` なので、全体で `O(N log N)` です。空間計算量は、仕事の配列を持つので `O(N)` です。

```python
import sys

def main():
    input = sys.stdin.readline

    N = int(input())

    jobs = []
    for _ in range(N):
        d, t = map(int, input().split())
        jobs.append((t, d))

    jobs.sort()

    now = 0

    for t, d in jobs:
        now += d
        if now > t:
            print("No")
            return

    print("Yes")

if __name__ == "__main__":
    main()
```

この問題でよくあるミスは、所要時間 `d_i` が短い順に処理してしまうことです。短い仕事から処理するのは一見よさそうですが、締め切りが厳しい長い仕事を後回しにすると失敗することがあります。

たとえば、

```text
仕事A: d = 1, t = 100
仕事B: d = 10, t = 10
```

短い順にすると `A → B` ですが、仕事B の完了時刻は `11` になってしまい、締め切り `10` に間に合いません。

一方、締め切り順に `B → A` とすれば、仕事B は時刻 `10` に終わり、仕事A は時刻 `11` に終わります。どちらも間に合います。

したがって、この問題では「短い順」ではなく、「締め切りが早い順」が本質です。
