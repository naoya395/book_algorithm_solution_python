## 0. ヒープで何を保証したいか

今回は **最大値ヒープ** を扱います。

最大値ヒープでは、すべての親について、

```python
親の値 >= 子の値
```

が成り立ちます。

そのため、最大値は必ず `heap[0]` にあります。

例：

```python
heap = [10, 7, 8, 3, 5, 1]
```

これは最大値ヒープです。`10` が根で、どの親も子以上になっています。

````python
# 補助：配列をヒープの段ごとに表示するだけの関数です。
# 実装課題ではありません。

def show_heap(heap):
    start = 0
    width = 1
    level = 0
    while start < len(heap):
        print(f"level {level}:", heap[start:start + width])
        start += width
        width *= 2
        level += 1

show_heap([10, 7, 8, 3, 5, 1])
````

## 16. おまけ：標準ライブラリ heapq との違い

Python 標準の `heapq` は **最小値ヒープ** です。

そのため、最大値ヒープとして使いたいときは、値にマイナスを付けて入れることがよくあります。

```python
import heapq

heapq.heappush(heap, -x)
max_value = -heapq.heappop(heap)
```

ここは自作ヒープの理解が終わってから見れば十分です。

````python
# おまけ：heapq を最大値ヒープのように使う例

import heapq

def descending_order_with_heapq(nums):
    h = []
    for x in nums:
        heapq.heappush(h, -x)

    ans = []
    while h:
        ans.append(-heapq.heappop(h))
    return ans

assert descending_order_with_heapq([5, 3, 7, 1]) == [7, 5, 3, 1]

print("OK: heapq example")
````
