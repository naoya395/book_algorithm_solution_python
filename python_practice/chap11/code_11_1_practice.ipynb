## 5. サイズが大きい方を根にする

何も考えずに併合すると、木が深くなって `find` が遅くなることがあります。

そこで、サイズが大きいグループの根を新しい根にします。  
これをすると、木が深くなりにくくなります。

### 問題

`union(x, y)` の中で、`size_list[x] < size_list[y]` のとき `x` と `y` を入れ替えてください。

````python
class UnionFind:
    def __init__(self, n):
        self.parent = [-1] * n
        self.size_list = [1] * n

    def find(self, x):
        if self.parent[x] == -1:
            return x
        return self.find(self.parent[x])

    def same(self, x, y):
        return self.find(x) == self.find(y)

    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return False

        # TODO: サイズが大きい方を x にする

        self.parent[y] = x
        self.size_list[x] += self.size_list[y]

        return True

    def size(self, x):
        return self.size_list[self.find(x)]
````

### テスト

````python
uf = UnionFind(6)

uf.union(0, 1)  # サイズ2
uf.union(2, 3)  # サイズ2
uf.union(2, 4)  # サイズ3

uf.union(0, 2)  # 大きい方である 2 側が根になるはず

root = uf.find(2)

assert uf.find(0) == root
assert uf.find(1) == root
assert uf.find(3) == root
assert uf.find(4) == root
assert uf.size(0) == 5
assert uf.same(1, 4) == True
assert uf.same(0, 5) == False

print("OK")
````

### 模範解答

````python
class UnionFind:
    def __init__(self, n):
        self.parent = [-1] * n
        self.size_list = [1] * n

    def find(self, x):
        if self.parent[x] == -1:
            return x
        return self.find(self.parent[x])

    def same(self, x, y):
        return self.find(x) == self.find(y)

    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return False

        if self.size_list[x] < self.size_list[y]:
            x, y = y, x

        self.parent[y] = x
        self.size_list[x] += self.size_list[y]

        return True

    def size(self, x):
        return self.size_list[self.find(x)]
````
