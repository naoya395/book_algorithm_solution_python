## 1. アルゴリズムとコードの詳しい解説

```python
INF = 20000000


def main() -> None:
    n = int(input())
    a = list(map(int, input().split()))

    min_value = INF
    for i in range(n):
        if a[i] < min_value:
            min_value = a[i]

    print(min_value)


if __name__ == "__main__":
    main()
```

### 目的

十分大きい初期値 `INF` から始め、配列 `a` の各要素と現在の最小値 `min_value` を比較して更新します。

### 変数・関数の意味

- `INF`: 最小値探索の初期値として使う十分大きな数です。
- `n`: 配列の要素数です。
- `a`: 最小値を探す配列です。
- `min_value`: 現時点で見つかっている最小値です。
- `i`: 現在見ている添字です。

### 処理の流れ

1. `n` と配列 `a` を読み込みます。
2. `min_value = INF` とします。
3. 全要素 `a[i]` を順に見ます。
4. `a[i] < min_value` なら `min_value` を `a[i]` に更新します。
5. 最後に `min_value` を出力します。

### 計算量

全要素を 1 回ずつ見るため時間計算量は **O(n)**、追加メモリは **O(1)** です。
