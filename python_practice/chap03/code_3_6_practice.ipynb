`n` 個の要素を選ぶ／選ばないの全パターンをビットで表し、和が `w` になる部分集合があるか判定します。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n, w = map(int, input().split())
    a = list(map(int, input().split()))

    exist = False
    for bit in range(1 << n):
        total = 0
        for i in range(n):
            if bit & (1 << i):
                total += a[i]

        if total == w:
            exist = True

    if exist:
        print("Yes")
    else:
        print("No")


if __name__ == "__main__":
    main()
```

### 目的

各要素について「選ぶ」「選ばない」の 2 通りがあるため、全体で `2^n` 通りを調べます。整数 `bit` の各ビットを見て、選ばれた要素の合計 `total` を作ります。

### 変数・関数の意味

- `n`: 要素数です。
- `w`: 作りたい合計値です。
- `a`: 候補となる整数列です。
- `exist`: 合計が `w` になる選び方があるかを表します。
- `bit`: 選び方を表す整数です。
- `total`: 現在の選び方での合計です。
- `i`: どの要素を選ぶか確認する添字です。

### 処理の流れ

1. 入力を読み込みます。
2. `bit` を `0` から `2^n - 1` まで動かします。
3. 各 `bit` について `total = 0` から始めます。
4. `i` 番目のビットが立っていれば `a[i]` を足します。
5. `total == w` なら `exist = True` にします。
6. 最後に `Yes` または `No` を出力します。

### 計算量

選び方が `2^n` 通りあり、各選び方で最大 `n` 個を見るため **O(n 2^n)** です。


3,6 問題名：目標の重さ

N 個のおもりがあります。
i 番目のおもりの重さは A_i です。

これらのおもりの中からいくつかを選び、選んだおもりの重さの合計をちょうど W にできるか判定してください。

おもりは 1 個も選ばなくてもよいものとします。

制約

1 ≤ N ≤ 20
0 ≤ W ≤ 10^9
0 ≤ A_i ≤ 10^9
入力はすべて整数である。

入力

入力は以下の形式で標準入力から与えられる。

```
N W
A_1 A_2 ... A_N
```

出力

選んだおもりの重さの合計をちょうど W にできるなら `Yes` を、できないなら `No` を出力してください。

入力例 1

```
4 10
1 3 4 6
```

出力例 1

```
Yes
```

説明

重さ 4 のおもりと重さ 6 のおもりを選ぶと、合計は 10 になります。

入力例 2

```
3 11
2 5 8
```

出力例 2

```
No
```

説明

どのおもりの組み合わせを選んでも、合計を 11 にすることはできません。

想定解法

各おもりについて「選ぶ」または「選ばない」の 2 通りを考えると、全体で 2^N 通りの選び方があります。
すべての選び方について合計を計算し、その中に W と等しいものがあるかを調べればよいです。

計算量は O(N 2^N) です。
N ≤ 20 なので、この方法で十分間に合います。
