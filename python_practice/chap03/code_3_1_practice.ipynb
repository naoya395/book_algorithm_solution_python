配列 `a` の中に目的の値 `v` が存在するかを、先頭から順に調べる線形探索を練習します。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n, v = map(int, input().split())
    a = list(map(int, input().split()))

    exist = False
    for i in range(n):
        if a[i] == v:
            exist = True

    if exist:
        print("Yes")
    else:
        print("No")


if __name__ == "__main__":
    main()
```

### 目的

`n` 個の整数からなる配列 `a` に、値 `v` が含まれているかを判定します。見つかったら `exist` を `True` にし、最後に `Yes` または `No` を出力します。

### 変数・関数の意味

- `n`: 配列の要素数です。
- `v`: 探したい値です。
- `a`: 探索対象の整数列です。
- `exist`: 値 `v` が見つかったかどうかを表す真偽値です。最初は `False` にします。
- `i`: 現在調べている添字です。

### 処理の流れ

1. `n` と `v` を入力します。
2. 配列 `a` を入力します。
3. `exist = False` として、まだ見つかっていない状態から始めます。
4. `i = 0` から `n - 1` まで順番に調べます。
5. `a[i] == v` なら `exist = True` にします。
6. 最後に `exist` が真なら `Yes`、偽なら `No` を出力します。

### 計算量

配列を最大 `n` 個調べるので時間計算量は **O(n)**、追加で使うメモリは **O(1)** です。
