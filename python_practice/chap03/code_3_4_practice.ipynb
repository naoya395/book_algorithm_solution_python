# code_3_4 演習ノート：2 つの配列から条件を満たす最小和を探す

元コード：`python_codes/chap03/code_3_4.py`

2 重ループで `a[i] + b[j]` をすべて調べ、`k` 以上になる組の最小値を求めます。


## 1. アルゴリズムとコードの詳しい解説

```python
INF = 20000000


def main() -> None:
    n, k = map(int, input().split())
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))

    min_value = INF
    for i in range(n):
        for j in range(n):
            if a[i] + b[j] < k:
                continue

            if a[i] + b[j] < min_value:
                min_value = a[i] + b[j]

    print(min_value)


if __name__ == "__main__":
    main()
```

### 目的

配列 `a` と `b` から 1 つずつ選んだ和のうち、`k` 以上で最小のものを全探索で求めます。条件を満たさない組は `continue` で飛ばします。

### 変数・関数の意味

- `INF`: 初期値として使う大きな数です。
- `n`: 各配列の要素数です。
- `k`: 和が満たすべき下限です。
- `a`, `b`: 候補値の配列です。
- `min_value`: 条件を満たす和の最小値です。
- `i`, `j`: `a` と `b` の添字です。

### 処理の流れ

1. 入力を読み込みます。
2. `min_value = INF` とします。
3. `a[i]` と `b[j]` の全組合せを 2 重ループで調べます。
4. 和が `k` 未満なら候補にできないため `continue` します。
5. 和が `min_value` より小さければ更新します。
6. 最後に `min_value` を出力します。

### 計算量

全組合せ `n × n` を調べるので時間計算量は **O(n²)** です。
