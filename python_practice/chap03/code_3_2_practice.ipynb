元コード：`python_codes/chap03/code_3_2.py`

配列 `a` から値 `v` を探し、最初に見つかった添字を出力する線形探索を練習します。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n, v = map(int, input().split())
    a = list(map(int, input().split()))

    found_id = -1
    for i in range(n):
        if a[i] == v:
            found_id = i
            break

    print(found_id)


if __name__ == "__main__":
    main()
```

### 目的

値 `v` が最初に現れる位置を探します。見つからなかった場合は `-1` を出力します。見つかった時点で `break` するため、不要な探索を止められます。

### 変数・関数の意味

- `n`: 配列の要素数です。
- `v`: 探す値です。
- `a`: 探索対象の配列です。
- `found_id`: 見つかった添字です。見つかっていないことを表すため初期値を `-1` にします。
- `i`: 現在調べている添字です。

### 処理の流れ

1. 入力を読み込みます。
2. `found_id = -1` として探索を始めます。
3. `a[i] == v` なら `found_id = i` とします。
4. 最初の位置だけが必要なので `break` でループを終了します。
5. `found_id` を出力します。

### 計算量

最悪の場合は全要素を調べるため **O(n)**、追加メモリは **O(1)** です。
