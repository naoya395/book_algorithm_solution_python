### 答え

````python
def total_sum_with_plus(S: str) -> int:
    N = len(S)
    answer = 0

    for bit in range(1 << (N - 1)):
        current = 0

        for i in range(N):
            current = current * 10 + int(S[i])

            # 最後の文字に到達した場合、または i 番目の後ろに + を入れる場合
            if i == N - 1 or bit & (1 << i):
                answer += current
                current = 0

    return answer


S = input().strip()
print(total_sum_with_plus(S))
````
