# code_2_2 演習ノート：2 重ループと計算量 O(n²)

このノートでは、同じ範囲を 2 重に走査するコードを書けるようになり、計算量が **O(n²)** になる理由を理解します。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n = int(input())

    count = 0
    for _ in range(n):
        for _ in range(n):
            count += 1


if __name__ == "__main__":
    main()
```

### 変数の意味

- `n`
  - 標準入力から読み込む整数です。
  - 外側のループも内側のループも、この `n` 回ずつ繰り返します。
- `count`
  - 内側ループの処理が何回実行されたかを数える変数です。
  - 最初は `0` です。
  - 一番内側で `count += 1` を実行します。
- 外側の `_`
  - 外側ループの回数を表します。
  - 値そのものは使わないため `_` にしています。
- 内側の `_`
  - 内側ループの回数を表します。
  - こちらも値そのものを使わないため `_` にしています。

### 処理の流れ

1. `n` を入力します。
2. `count` を `0` にします。
3. 外側ループを `n` 回実行します。
4. 外側ループ 1 回につき、内側ループを `n` 回実行します。
5. 内側ループのたびに `count` を 1 増やします。

たとえば `n = 3` のとき、外側ループは 3 回です。各外側ループの中で内側ループが 3 回動くので、`count += 1` は `3 × 3 = 9` 回実行されます。

### 計算量

内側の処理は合計で `n × n` 回実行されます。したがって時間計算量は **O(n²)** です。
