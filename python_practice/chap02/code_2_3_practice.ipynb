## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n = int(input())

    for i in range(2, n + 1, 2):
        print(i)


if __name__ == "__main__":
    main()
```

### 変数の意味

- `n`
  - 標準入力から読み込む整数です。
  - `n` 以下の偶数を出力するための上限になります。
- `i`
  - 現在出力しようとしている偶数です。
  - `range(2, n + 1, 2)` によって、`2, 4, 6, ...` と 2 ずつ増えます。

### `range(2, n + 1, 2)` の意味

Python の `range(start, stop, step)` は、`start` から始めて、`stop` 未満まで、`step` ずつ増える整数列を作ります。

- `start = 2`
  - 正の偶数の最初の値である `2` から始めます。
- `stop = n + 1`
  - `range` は `stop` を含まないので、`n` 自身を含めたい場合は `n + 1` にします。
- `step = 2`
  - 偶数だけをたどるために、2 ずつ増やします。

### 処理の流れ

1. `n` を入力します。
2. `i` を `2` から始めます。
3. `i` が `n` 以下である間、`i` を出力します。
4. 次の `i` は 2 増えます。

たとえば `n = 7` なら、出力は `2`, `4`, `6` です。`8` は `7` を超えるので出力されません。

### 計算量

出力される偶数の個数はおよそ `n / 2` 個です。定数倍は計算量の表記では省略するため、時間計算量は **O(n)** です。
