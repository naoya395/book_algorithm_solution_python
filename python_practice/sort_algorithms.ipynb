## 解答

```python
def merge_sort(a, left=0, right=None):
    if right is None:
        right = len(a)

    # 要素数が 0 個または 1 個なら、もうソート済み
    if right - left <= 1:
        return

    mid = (left + right) // 2

    # 左半分と右半分をそれぞれソート
    merge_sort(a, left, mid)
    merge_sort(a, mid, right)

    # 2つのソート済み配列を合体する
    buf = []
    i = left
    j = mid

    while i < mid and j < right:
        if a[i] <= a[j]:
            buf.append(a[i])
            i += 1
        else:
            buf.append(a[j])
            j += 1

    # 左側が余った場合
    while i < mid:
        buf.append(a[i])
        i += 1

    # 右側が余った場合
    while j < right:
        buf.append(a[j])
        j += 1

    # 元の配列に戻す
    for k in range(len(buf)):
        a[left + k] = buf[k]
```

```python
a = [5, 2, 4, 6, 1, 3]
merge_sort(a)
print(a)
```

```python
def quick_sort(a, left=0, right=None):
    if right is None:
        right = len(a)

    # 要素数が 0 個または 1 個なら、もうソート済み
    if right - left <= 1:
        return

    # pivot を選ぶ。ここでは真ん中の値を使う
    pivot_index = (left + right) // 2
    pivot = a[pivot_index]

    # pivot を一旦右端に移動する
    a[pivot_index], a[right - 1] = a[right - 1], a[pivot_index]

    # i は「pivot 未満の値を置く場所」
    i = left

    for j in range(left, right - 1):
        if a[j] < pivot:
            a[i], a[j] = a[j], a[i]
            i += 1

    # pivot を正しい位置に置く
    a[i], a[right - 1] = a[right - 1], a[i]

    # pivot の左側と右側を再帰的にソート
    quick_sort(a, left, i)
    quick_sort(a, i + 1, right)
```

```python
def bucket_sort(a, max_value):
    num = [0] * (max_value + 1)

    # 各値の個数を数える
    for x in a:
        num[x] += 1

    # num[v] = v 以下の個数、になるように累積和を取る
    for v in range(1, max_value + 1):
        num[v] += num[v - 1]

    # ソート結果を入れる配列
    b = [0] * len(a)

    # 後ろから見ることで安定ソートになる
    for i in range(len(a) - 1, -1, -1):
        x = a[i]
        num[x] -= 1
        b[num[x]] = x

    # 元の配列に戻す
    for i in range(len(a)):
        a[i] = b[i]
```
