# 5.1 Vacation

あなたはこれから `N` 日間の夏休みを過ごします。

各日には、次の 3 種類の行動のうちちょうど 1 つを選びます。

1. 海で泳ぐ
2. 虫捕りをする
3. 宿題をする

`i` 日目にそれぞれの行動をしたときに得られる幸福度は、順に `a_i, b_i, c_i` です。

ただし、2 日連続で同じ種類の行動を選ぶことはできません。

`N` 日間で得られる幸福度の合計の最大値を求めてください。

## 入力

```text
N
a_1 b_1 c_1
a_2 b_2 c_2
...
a_N b_N c_N
```

## 出力

`N` 日間で得られる幸福度の合計の最大値を出力してください。

## 制約

```text
1 ≤ N ≤ 100000
1 ≤ a_i, b_i, c_i ≤ 10000
入力はすべて整数
```

## 入力例

```text
3
10 40 70
20 50 80
30 60 90
```

## 出力例

```text
210
```

例えば、1 日目に宿題、2 日目に虫捕り、3 日目に宿題を選ぶと、幸福度の合計は `70 + 50 + 90 = 210` になります。

# 5.2 部分和問題

`N` 個の正の整数 `a_0, a_1, ..., a_{N-1}` と、正の整数 `W` が与えられます。

これらの整数の中からいくつかを選び、その総和をちょうど `W` にできるかどうかを判定してください。

ただし、それぞれの整数は高々 1 回だけ選ぶことができます。整数を 1 つも選ばないことも許されます。

## 入力

```text
N W
a_0 a_1 ... a_{N-1}
```

## 出力

総和をちょうど `W` にできるなら `Yes`、できないなら `No` と出力してください。

## 制約

```text
1 ≤ N ≤ 100
1 ≤ W ≤ 100000
1 ≤ a_i ≤ W
入力はすべて整数
```

## 入力例 1

```text
4 11
2 5 9 10
```

## 出力例 1

```text
Yes
```

`2` と `9` を選ぶと、総和を `11` にできます。

## 入力例 2

```text
3 8
2 4 6
```

## 出力例 2

```text
No
```

どのように選んでも総和を `8` にすることはできません。

# 5.3 作れる総和の種類数

`N` 個の正の整数 `a_0, a_1, ..., a_{N-1}` と、正の整数 `W` が与えられます。

これらの整数の中からいくつかを選び、その総和として作ることができる整数を考えます。

`1` 以上 `W` 以下の整数のうち、総和として作ることができるものが何種類あるかを求めてください。

ただし、それぞれの整数は高々 1 回だけ選ぶことができます。整数を 1 つも選ばないことも許されますが、その場合の総和 `0` は答えには数えません。

また、数えるのは「作れる整数の種類数」です。同じ整数を作る選び方が複数あっても、その整数は 1 回だけ数えます。

## 入力

```text
N W
a_0 a_1 ... a_{N-1}
```

## 出力

`1` 以上 `W` 以下の整数のうち、作ることができる整数の個数を出力してください。

## 制約

```text
1 ≤ N ≤ 100
1 ≤ W ≤ 100000
1 ≤ a_i ≤ W
入力はすべて整数
```

## 入力例

```text
3 10
2 3 5
```

## 出力例

```text
6
```

作ることができる `1` 以上 `10` 以下の整数は、`2, 3, 5, 7, 8, 10` の 6 種類です。

例えば `5` は、`5` を選ぶ方法と、`2` と `3` を選ぶ方法の 2 通りで作れますが、整数 `5` としては 1 回だけ数えます。

# 5.4 K 個以内で W を作る問題

`N` 個の正の整数 `a_0, a_1, ..., a_{N-1}` と、正の整数 `K, W` が与えられます。

これらの整数の中から `K` 個以下を選び、その総和をちょうど `W` にできるかどうかを判定してください。

ただし、それぞれの整数は高々 1 回だけ選ぶことができます。

## 入力

```text
N K W
a_0 a_1 ... a_{N-1}
```

## 出力

`K` 個以下の整数を選んで総和をちょうど `W` にできるなら `Yes`、できないなら `No` と出力してください。

## 制約

```text
1 ≤ N ≤ 100
1 ≤ K ≤ N
1 ≤ W ≤ 100000
1 ≤ a_i ≤ W
入力はすべて整数
```

## 入力例 1

```text
5 2 10
1 3 4 6 8
```

## 出力例 1

```text
Yes
```

`4` と `6` を選ぶと、2 個以内で総和を `10` にできます。

## 入力例 2

```text
5 1 10
1 3 4 6 8
```

## 出力例 2

```text
No
```

1 個だけ選んで総和を `10` にすることはできません。
