## 5.1 Vacation

あなたはこれから `N` 日間の夏休みを過ごします。

各日には、次の 3 種類の行動のうちちょうど 1 つを選びます。

1. 海で泳ぐ
2. 虫捕りをする
3. 宿題をする

`i` 日目にそれぞれの行動をしたときに得られる幸福度は、順に `a_i, b_i, c_i` です。

ただし、2 日連続で同じ種類の行動を選ぶことはできません。

`N` 日間で得られる幸福度の合計の最大値を求めてください。

## 入力

```text
N
a_1 b_1 c_1
a_2 b_2 c_2
...
a_N b_N c_N
```

## 出力

`N` 日間で得られる幸福度の合計の最大値を出力してください。

## 制約

```text
1 ≤ N ≤ 100000
1 ≤ a_i, b_i, c_i ≤ 10000
入力はすべて整数
```

## 入力例

```text
3
10 40 70
20 50 80
30 60 90
```

## 出力例

```text
210
```

例えば、1 日目に宿題、2 日目に虫捕り、3 日目に宿題を選ぶと、幸福度の合計は `70 + 50 + 90 = 210` になります。

# 5.1 - 5.4 解説



## 5.1 Vacation

各日に 3 つの行動のうち 1 つを選びます。ただし、前日と同じ行動は選べません。

何も制約がなければ、各日について一番幸福度が高い行動を選べばよいです。しかし今回は「2 日連続で同じ行動をしてはいけない」ので、ある日だけを独立に考えることはできません。

そこで、「最後にどの行動をしたか」を状態として持ちます。

`dp[i][j]` を次のように定義します。

```text
dp[i][j] = i 日目までの行動を決めたとき、
           i 日目に行動 j を選んだ場合の幸福度の最大値
```

ここで、行動 `j` は次のように対応させます。

```text
j = 0 : 海で泳ぐ
j = 1 : 虫捕りをする
j = 2 : 宿題をする
```

`i` 日目に行動 `j` を選ぶには、`i-1` 日目には `j` 以外の行動を選んでいる必要があります。

したがって、遷移は次のようになります。

```text
dp[i][j] = max(dp[i-1][k]) + value[i][j]
ただし k != j
```

つまり、前日に選んだ行動 `k` が今日の行動 `j` と異なるものだけを候補にします。

初期条件は、1 日目について直接決めればよいです。

```text
dp[0][0] = a_0
dp[0][1] = b_0
dp[0][2] = c_0
```

最後に、`N` 日目の行動はどれでもよいので、

```text
max(dp[N-1][0], dp[N-1][1], dp[N-1][2])
```

が答えです。

状態数は `N × 3` 個で、それぞれの状態について高々 3 通りを見るだけなので、計算量は `O(N)` です。

Python コードは次のようになります。

```python
N = int(input())

value = []
for _ in range(N):
    a, b, c = map(int, input().split())
    value.append([a, b, c])

dp = [[0] * 3 for _ in range(N)]

# 初期条件：0 日目にそれぞれの行動を選んだ場合
for j in range(3):
    dp[0][j] = value[0][j]

# DP
for i in range(1, N):
    for j in range(3):
        best = 0
        for k in range(3):
            if j == k:
                continue
            best = max(best, dp[i - 1][k])
        dp[i][j] = best + value[i][j]

print(max(dp[N - 1]))
```

## 5.2 部分和問題

`N` 個の整数からいくつか選んで、総和をちょうど `W` にできるかを判定する問題です。

各整数は「選ぶ」か「選ばない」かの 2 択です。ただし、すべての選び方を試すと `2^N` 通りになってしまい、`N` が大きいと間に合いません。

そこで、「ここまで見た整数で、ある総和を作れるか」を DP で管理します。

`dp[i][j]` を次のように定義します。

```text
dp[i][j] = 最初の i 個の整数からいくつか選んで、
           総和を j にできるか
```

値は `True` または `False` です。

次に `a_i` を見るとき、選択肢は 2 つあります。

まず、`a_i` を選ばない場合です。

すでに総和 `j` を作れているなら、`a_i` を使わなくても総和 `j` は作れます。

```text
dp[i][j] が True なら dp[i+1][j] も True
```

次に、`a_i` を選ぶ場合です。

総和 `j` を作りたいなら、`a_i` を選ぶ前には総和 `j - a_i` を作れていればよいです。

```text
j >= a_i かつ dp[i][j-a_i] が True なら dp[i+1][j] も True
```

初期条件は、

```text
dp[0][0] = True
```

です。これは「何も選ばなければ総和 0 を作れる」という意味です。

最終的に、すべての整数を見たあとに総和 `W` を作れるかを見ればよいので、答えは `dp[N][W]` です。

計算量は `O(NW)` です。

実装では、2 次元 DP をそのまま使う代わりに、1 次元配列に圧縮できます。

ただし、各整数を高々 1 回しか使えないので、`j` は大きい方から小さい方へ更新します。小さい方から更新すると、同じ `a_i` を何度も使ったことになってしまうためです。

Python コードは次のようになります。

```python
N, W = map(int, input().split())
a = list(map(int, input().split()))

dp = [[False] * (W + 1) for _ in range(N + 1)]
dp[0][0] = True

for i in range(N):
    for j in range(W + 1):
        if not dp[i][j]:
            continue

        # a[i] を選ばない
        dp[i + 1][j] = True

        # a[i] を選ぶ
        if j + a[i] <= W:
            dp[i + 1][j + a[i]] = True

if dp[N][W]:
    print("Yes")
else:
    print("No")
```

## 5.3 作れる総和の種類数

この問題は、5.2 の部分和問題とほとんど同じです。

違いは、5.2 では「総和 `W` を作れるか」だけを判定すればよかったのに対し、この問題では「`1` 以上 `W` 以下のうち、作れる整数が何種類あるか」を数える点です。

ただし、注意すべき点があります。

ここで数えるのは「選び方の数」ではありません。「作れる総和の種類数」です。

例えば、整数列が

```text
2 3 5
```

のとき、総和 `5` は、

```text
5 を選ぶ
2 と 3 を選ぶ
```

の 2 通りで作れます。

しかし、答えとして数えるときは、整数 `5` は 1 回だけ数えます。

そのため、DP は「何通りあるか」ではなく、「作れるかどうか」を管理すれば十分です。

`dp[i][j]` を次のように定義します。

```text
dp[i][j] = 最初の i 個の整数からいくつか選んで、
           総和 j を作れるか
```

これは 5.2 と同じです。

遷移も同じです。

`a_i` を選ばない場合は、すでに作れる総和をそのまま引き継ぎます。

`a_i` を選ぶ場合は、`j - a_i` が作れていれば、そこに `a_i` を足して `j` を作れます。

初期条件は、

```text
dp[0][0] = True
```

です。

最後に、`1` から `W` までの整数について、作れるものを数えます。

```text
答え = dp[N][1], dp[N][2], ..., dp[N][W] のうち True の個数
```

`0` は「何も選ばない」ことで常に作れてしまいますが、この問題では `1` 以上 `W` 以下だけを数えるので、`0` は答えに含めません。

計算量は `O(NW)` です。

Python コードは次のようになります。

```python
N, W = map(int, input().split())
a = list(map(int, input().split()))

dp = [False] * (W + 1)
dp[0] = True

for x in a:
    for j in range(W, x - 1, -1):
        if dp[j - x]:
            dp[j] = True

ans = 0
for j in range(1, W + 1):
    if dp[j]:
        ans += 1

print(ans)
```

これは俺が頑張って書いた，２次元配列を使う解答．

## 5.4 K 個以内で W を作る問題

この問題では、N 個の正の整数から K 個以下を選び、その総和をちょうど W にできるかどうかを判定します。

ただし、それぞれの整数は高々 1 回しか選ぶことができません。

## 考え方

普通に考えると、

```text
dp[i][j][k] = 最初の i 個の整数から k 個選んで、総和 j を作れるか
```

のように、選んだ個数まで状態に持つ方法が考えられます。

しかし、この方法だと状態数が N × W × K になり、計算量は O(NWK) になります。

この問題で知りたいのは、「ちょうど K 個選べるか」ではなく、「K 個以下で作れるか」です。

つまり、総和 j を作る方法のうち、選ぶ個数の最小値が分かれば十分です。

そこで、DP の値を True / False ではなく、「その総和を作るために必要な最小個数」として持ちます。

## 2 次元 DP で考える

まずは、次のような DP を考えます。

```text
dp[i][j] = 最初の i 個の整数からいくつか選んで、
           総和 j を作るために必要な個数の最小値
```

もし総和 j を作れない場合は、十分大きい値 INF を入れておきます。

初期条件は、

```text
dp[0][0] = 0
```

です。

何も選ばなければ、総和 0 を 0 個で作れるからです。

それ以外の `dp[0][j]` は作れないので INF です。

次に、i 番目の整数を `a_i` として、遷移を考えます。

まず、`a_i` を選ばない場合です。

この場合、総和 j を作るために必要な個数は変わりません。

```text
dp[i + 1][j] = min(dp[i + 1][j], dp[i][j])
```

次に、`a_i` を選ぶ場合です。

総和 j を作りたいなら、その前に総和 `j - a_i` を作っておき、そこに `a_i` を追加します。

このとき、選ぶ個数は 1 増えます。

```text
dp[i + 1][j] = min(dp[i + 1][j], dp[i][j - a_i] + 1)
```

したがって、最終的に

```text
dp[N][W] <= K
```

なら、K 個以下の整数を選んで総和 W を作れるので `Yes` です。

そうでなければ `No` です。

## なぜ最小個数を持てばよいか

この問題の目的は、「総和 W を K 個以下で作れるかどうか」を判定することです。

そのため、総和 W を作る方法が複数ある場合でも、その中で最も少ない個数だけ分かれば十分です。

もし総和 W を作るための最小個数が K 以下なら、条件を満たす選び方が存在します。

逆に、最小個数ですら K より大きいなら、どの選び方でも K 個以内にはできません。

したがって、各総和について「作れるかどうか」だけではなく、「作るために必要な最小個数」を持つことで、この問題を解くことができます。

## 1 次元 DP への圧縮

ここまでの説明では、`dp[i][j]` という 2 次元 DP で考えました。

しかし、遷移を見ると、`dp[i + 1][j]` を計算するときに必要なのは、直前の状態である `dp[i][...]` だけです。

つまり、すべての i について DP テーブルを持っておく必要はありません。

そこで、実装では次のような 1 次元 DP に圧縮します。

```text
dp[j] = 今まで見た整数からいくつか選んで、
        総和 j を作るために必要な個数の最小値
```

初期状態は、

```text
dp[0] = 0
```

です。

これは、何も選ばなければ総和 0 を 0 個で作れるという意味です。

各整数 `x` について、`x` を使って総和 `j` を作る場合を考えると、

```text
dp[j] = min(dp[j], dp[j - x] + 1)
```

と更新できます。

ここで重要なのは、`j` を大きい方から小さい方へ更新することです。

なぜなら、この問題ではそれぞれの整数を高々 1 回しか使えないからです。

もし `j` を小さい方から大きい方へ更新してしまうと、同じ整数 `x` を使って更新した結果を、さらに同じ `x` で再び使ってしまう可能性があります。

それは、1 つの整数を何回も使うことになってしまうため、この問題の条件に合いません。

そのため、各 `x` について、

```text
j = W, W - 1, ..., x
```

の順に更新します。

## 実装

```python
N, K, W = map(int, input().split())
a = list(map(int, input().split()))

INF = 10**18

dp = [INF] * (W + 1)
dp[0] = 0

for x in a:
    for j in range(W, x - 1, -1):
        dp[j] = min(dp[j], dp[j - x] + 1)

if dp[W] <= K:
    print("Yes")
else:
    print("No")
```

## 計算量

配列 `dp` の大きさは `W + 1` です。

各整数について、総和 `0` から `W` までを考えるので、計算量は

```text
O(NW)
```

です。

また、1 次元 DP に圧縮しているため、メモリ使用量は

```text
O(W)
```

です。
