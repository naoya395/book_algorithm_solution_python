## 1. アルゴリズムとコードの詳しい解説



### 目的

文字列 `s` を `t` に変える最小操作回数を DP で求めます。置換・削除・挿入を遷移として考えます。

### 変数・関数の意味

- `s`, `t`: 比較する文字列です。
- `dp[i][j]`: `s` の先頭 `i` 文字と `t` の先頭 `j` 文字の編集距離です。
- `INF`: 未更新を表す大きな値です。

### 処理の流れ

1. 一致または置換、削除、挿入の候補で `dp[i][j]` を最小化します。

### 計算量

時間計算量は **O(|s||t|)** です。




## 問題名：文字列の編集距離

英小文字からなる 2 つの文字列 `S` と `T` が与えられます。

あなたは文字列 `S` に対して、以下の操作を何回でも行うことができます。

1. 文字を 1 文字削除する
2. 好きな位置に文字を 1 文字挿入する
3. 文字を 1 文字別の文字に置き換える

文字列 `S` を文字列 `T` に一致させるために必要な操作回数の最小値を求めてください。

## 入力

```text
S T
```

## 出力

`S` を `T` に変換するために必要な操作回数の最小値を出力してください。

## 制約

```text
1 ≤ |S| ≤ 2000
1 ≤ |T| ≤ 2000
S, T は英小文字のみからなる文字列
```

## 入力例 1

```text
kitten sitting
```

## 出力例 1

```text
3
```

## 説明

例えば、次のように 3 回の操作で `kitten` を `sitting` に変換できます。

```text
kitten
sitten   k を s に置き換える
sittin   e を i に置き換える
sitting  g を末尾に挿入する
```

2 回以下の操作では変換できないため、答えは `3` です。

## 入力例 2

```text
abc abc
```

## 出力例 2

```text
0
```

## 説明

すでに 2 つの文字列は等しいため、操作は必要ありません。

## 入力例 3

```text
abc axc
```

## 出力例 3

```text
1
```

## 説明

`b` を `x` に置き換えることで、1 回の操作で一致させることができます。

## 3. 回答：問題1


````python
INF = 1 << 29


def main() -> None:
    s, t = input().split()

    dp = [[INF] * (len(t) + 1) for _ in range(len(s) + 1)]
    dp[0][0] = 0

    for i in range(len(s) + 1):
        for j in range(len(t) + 1):
            if i > 0 and j > 0:
                if s[i - 1] == t[j - 1]:
                    dp[i][j] = min(dp[i][j], dp[i - 1][j - 1])
                else:
                    dp[i][j] = min(dp[i][j], dp[i - 1][j - 1] + 1)

            if i > 0:
                dp[i][j] = min(dp[i][j], dp[i - 1][j] + 1)

            if j > 0:
                dp[i][j] = min(dp[i][j], dp[i][j - 1] + 1)

    print(dp[len(s)][len(t)])


if __name__ == "__main__":
    main()
````
