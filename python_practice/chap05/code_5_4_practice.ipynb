## 1. アルゴリズムとコードの詳しい解説

```python
INF = 1 << 60


def chmin(a: int, b: int) -> int:
    if a > b:
        return b
    return a


def main() -> None:
    n = int(input())
    h = list(map(int, input().split()))

    dp = [INF] * n
    dp[0] = 0

    for i in range(n):
        if i + 1 < n:
            dp[i + 1] = chmin(dp[i + 1], dp[i] + abs(h[i] - h[i + 1]))
        if i + 2 < n:
            dp[i + 2] = chmin(dp[i + 2], dp[i] + abs(h[i] - h[i + 2]))

    print(dp[n - 1])


if __name__ == "__main__":
    main()
```

### 目的

現在の足場 `i` から、次の足場 `i+1` と `i+2` へ最小コストを配ります。

### 変数・関数の意味

- `dp[i]`: 足場 `i` までの最小コストです。
- `i + 1`, `i + 2`: 遷移先の足場です。
- `chmin`: 遷移先の値を小さく更新します。

### 処理の流れ

1. `dp[0]=0` から始め、各足場から到達可能な先へ値を更新します。

### 計算量

時間計算量は **O(n)** です。
