## 1. アルゴリズムとコードの詳しい解説

```python
INF = 1 << 60


def main() -> None:
    n = int(input())
    c = [[0] * (n + 1) for _ in range(n + 1)]
    for i in range(n + 1):
        c[i] = list(map(int, input().split()))

    dp = [INF] * (n + 1)
    dp[0] = 0

    for i in range(n + 1):
        for j in range(i):
            dp[i] = min(dp[i], dp[j] + c[j][i])

    print(dp[n])


if __name__ == "__main__":
    main()
```

### 目的

地点 `0` から `n` まで進むとき、`j` から `i` へ移動するコスト `c[j][i]` を使って最小コストを求めます。

### 変数・関数の意味

- `c[j][i]`: `j` から `i` へ移動するコストです。
- `dp[i]`: 地点 `i` までの最小コストです。
- `j`: 直前の分割位置です。

### 処理の流れ

1. すべての `i` について、直前の位置 `j < i` を全探索して `dp[i]` を更新します。

### 計算量

時間計算量は **O(n²)** です。
