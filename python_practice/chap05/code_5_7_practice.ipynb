品物を選ぶ／選ばないを順に考え、重さ上限 `w` 以内で価値の最大値を求めます。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    n, w = map(int, input().split())
    weight = [0] * n
    value = [0] * n
    for i in range(n):
        weight[i], value[i] = map(int, input().split())

    dp = [[0] * (w + 1) for _ in range(n + 1)]

    for i in range(n):
      for weight_sum in range(w + 1):
        dp[i + 1][weight_sum] = dp[i][weight_sum]

        if weight_sum - weight[i] >= 0:
            dp[i + 1][weight_sum] = max(
                dp[i + 1][weight_sum],
                dp[i][weight_sum - weight[i]] + value[i],
            )

            dp[i + 1][weight_sum] = max(dp[i + 1][weight_sum], dp[i][weight_sum])

    print(dp[n][w])


if __name__ == "__main__":
    main()
```

### 目的

品物を選ぶ／選ばないを順に考え、重さ上限 `w` 以内で価値の最大値を求めます。

### 変数・関数の意味

- `weight[i]`, `value[i]`: 品物 `i` の重さと価値です。
- `dp[i][weight_sum]`: 先頭 `i` 個の品物で重さ合計 `weight_sum` 以下にしたときの最大価値です。
- `max`: 現在の値と候補値の大きい方を選び、最大値更新を行います。

### 処理の流れ

1. 各品物について、選ぶ場合と選ばない場合の 2 通りで DP テーブルを更新します。

### 計算量

時間計算量は **O(nW)** です。


## 問題文

あなたは `N` 個の品物を持っています。

品物 `i` の重さは `weight_i`、価値は `value_i` です。

あなたは、重さの合計が `W` 以下になるように、いくつかの品物を選んでリュックに入れます。

ただし、各品物は **選ぶか選ばないかのどちらか** であり、同じ品物を 2 回以上選ぶことはできません。

選んだ品物の価値の合計として考えられる最大値を求めてください。

## 実装条件

この問題は、動的計画法を用いて解いてください。

`dp[i][j]` を、

```text
最初の i 個の品物だけを使って、重さの合計を j 以下にしたときの価値の最大値
```

として考えてください。

品物を選ばない場合と、選ぶ場合の 2 通りを比較しながら、表 `dp` を更新してください。

## 入力

入力は以下の形式で与えられます。

```text
N W
weight_1 value_1
weight_2 value_2
...
weight_N value_N
```

## 出力

選んだ品物の価値の合計の最大値を出力してください。

## 制約

```text
1 <= N <= 100
1 <= W <= 100000
1 <= weight_i <= W
1 <= value_i <= 100000
入力はすべて整数
```

## 入力例 1

```text
3 8
3 30
4 50
5 60
```

## 出力例 1

```text
90
```

## 説明

品物 1 と品物 3 を選ぶと、重さの合計は

```text
3 + 5 = 8
```

となり、リュックの重さ制限 `8` 以下です。

このとき価値の合計は

```text
30 + 60 = 90
```

となります。

これが最大なので、答えは `90` です。


## 説明
