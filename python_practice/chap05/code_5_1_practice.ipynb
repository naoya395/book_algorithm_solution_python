カエルが足場 `i` に来る最小コスト `dp[i]` を、直前 1 個または 2 個前の足場からの遷移で求めます。


## 1. アルゴリズムとコードの詳しい解説

```python
INF = 1 << 60


def main() -> None:
    n = int(input())
    h = list(map(int, input().split()))

    dp = [INF] * n
    dp[0] = 0

    for i in range(1, n):
        if i == 1:
            dp[i] = abs(h[i] - h[i - 1])
        else:
            dp[i] = min(
                dp[i - 1] + abs(h[i] - h[i - 1]),
                dp[i - 2] + abs(h[i] - h[i - 2]),
            )

    print(dp[n - 1])


if __name__ == "__main__":
    main()
```

### 目的

カエルが足場 `i` に来る最小コスト `dp[i]` を、直前 1 個または 2 個前の足場からの遷移で求めます。

### 変数・関数の意味

- `INF`: 未到達を表す十分大きい値です。
- `n`: 足場数です。
- `h`: 各足場の高さです。
- `dp[i]`: 足場 `i` に到達する最小コストです。

### 処理の流れ

1. `dp[0]=0` から始め、`i=1` 以降について 1 歩前・2 歩前から来るコストの小さい方を保存します。

### 計算量

時間計算量は **O(n)** です。
