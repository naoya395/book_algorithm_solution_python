カエルが足場 `i` に来る最小コスト `dp[i]` を、直前 1 個または 2 個前の足場からの遷移で求めます。


## 1. アルゴリズムとコードの詳しい解説

```python
INF = 1 << 60


def main() -> None:
    n = int(input())
    h = list(map(int, input().split()))

    dp = [INF] * n
    dp[0] = 0

    for i in range(1, n):
        if i == 1:
            dp[i] = abs(h[i] - h[i - 1])
        else:
            dp[i] = min(
                dp[i - 1] + abs(h[i] - h[i - 1]),
                dp[i - 2] + abs(h[i] - h[i - 2]),
            )

    print(dp[n - 1])


if __name__ == "__main__":
    main()
```

### 目的

カエルが足場 `i` に来る最小コスト `dp[i]` を、直前 1 個または 2 個前の足場からの遷移で求めます。

### 変数・関数の意味

- `INF`: 未到達を表す十分大きい値です。
- `n`: 足場数です。
- `h`: 各足場の高さです。
- `dp[i]`: 足場 `i` に到達する最小コストです。

### 処理の流れ

1. `dp[0]=0` から始め、`i=1` 以降について 1 歩前・2 歩前から来るコストの小さい方を保存します。

### 計算量

時間計算量は **O(n)** です。




## 問題文

カエルくんは、横一列に並んだ `N` 個の足場を使って、足場 `1` から足場 `N` まで移動しようとしています。

足場 `i` の高さは `H_i` です。

カエルくんは、現在いる足場から次のどちらかの移動ができます。

* 1 つ先の足場へ移動する
* 2 つ先の足場へ移動する

ただし、足場 `i` から足場 `j` へ移動するとき、疲労度が

`|H_i - H_j|`

だけ増えます。

カエルくんが足場 `1` から足場 `N` まで移動するとき、疲労度の合計の最小値を求めてください。

## 入力

入力は以下の形式で与えられます。

```text
N
H_1 H_2 ... H_N
```

## 出力

疲労度の合計の最小値を出力してください。

## 制約

```text
2 <= N <= 100000
1 <= H_i <= 10000
入力はすべて整数
```

## 入力例 1

```text
4
10 30 40 20
```

## 出力例 1

```text
30
```

## 説明

足場 `1 -> 2 -> 4` と移動すると、

```text
|10 - 30| + |30 - 20| = 20 + 10 = 30
```

となります。

これが最小なので、答えは `30` です。

## 入力例 2

```text
6
30 10 60 10 60 50
```

## 出力例 2

```text
40
```

このコードでは、`dp[i]` を「足場 `i` に到達するまでの最小疲労度」として計算しています。Python の添字は 0 始まりなので、問題文の足場 `1` がコード上では `0` に対応します。
