### 目的

`rec(i)` が足場 `i` への最小コストを返します。`dp` に保存して再計算を防ぎます。

### 変数・関数の意味

- `rec(i)`: 足場 `i` への最小コストを返す関数です。
- `dp[i]`: 計算済みなら答え、未計算なら `INF` です。
- `h`: 足場の高さです。

### 処理の流れ

1. 計算済みなら即返し、`i==0` を基本ケースにします。1 歩前・2 歩前から来る候補の最小を保存して返します。

### 計算量

時間計算量は **O(n)** です。
