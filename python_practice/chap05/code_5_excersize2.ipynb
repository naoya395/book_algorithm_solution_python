## 5. 何度でも使える整数

N 個の正整数 `a_1, a_2, ..., a_N` と、目標値 `W` が与えられます。

各整数 `a_i` は **0 回以上、何回でも** 使うことができます。
使った整数の総和をちょうど `W` にできるか判定してください。

## 入力

```text
N W
a_1 a_2 ... a_N
```

## 出力

総和をちょうど `W` にできるなら `Yes`、できないなら `No` を出力してください。

## 制約

```text
1 ≤ N ≤ 100
1 ≤ W ≤ 100000
1 ≤ a_i ≤ W
```

この問題は、時間計算量 `O(NW)`、空間計算量 `O(W)` で解けることを想定しています。

## 入力例 1

```text
3 17
3 5 8
```

## 出力例 1

```text
Yes
```

`8 + 3 + 3 + 3 = 17` とできます。同じ整数を複数回使ってよいことに注意してください。

## 入力例 2

```text
2 7
4 6
```

## 出力例 2

```text
No
```

`4` と `6` を何回使っても、作れる総和は偶数です。したがって `7` は作れません。

# 解説と想定解法




## 5. 何度でも使える整数

`dp[s]` を、「これまで見た整数を使って、総和 `s` を作れるか」とします。

最初は何も使わなければ総和 `0` を作れるので、

```python
dp[0] = True
```

です。

整数 `x` を処理するとき、`s - x` を作れるなら、そこに `x` を 1 個追加して `s` を作れます。

ここで重要なのは、`s` を小さい方から更新することです。
小さい方から見ることで、同じ `x` を使って更新された状態を、さらにもう一度使えます。つまり、`x` を何回でも使えることになります。

## 想定解法

```python
N, W = map(int, input().split())
a = list(map(int, input().split()))

dp = [False] * (W + 1)
dp[0] = True

for x in a:
    for s in range(x, W + 1):
        if dp[s - x]:
            dp[s] = True

print("Yes" if dp[W] else "No")
```

時間計算量は `O(NW)`、空間計算量は `O(W)` です。

# 6. 回数制限つき整数

5 では各整数を何回でも使えましたが、今回は `a_i` を最大 `m_i` 回までしか使えません。

単純に `a_i` を `m_i` 個に分解して 0-1 部分和問題にすると、`m_i` が大きいと間に合いません。
そのため、`O(NW)` で処理できる DP を使います。

`dp[s]` を次のように定義します。

```text
現在の整数 a_i まで見たとき、
総和 s を作れるなら、その状態からさらに使える a_i の残り個数
作れないなら -1
```

整数 `x = a_i`, 使用上限 `c = m_i` を処理するとき、総和 `s` について次の 3 通りを考えます。

まず、`s` が前の整数までで既に作れていたなら、今回の整数 `x` をまだ 1 個も使っていない状態にできます。
このとき、残り個数は `c` です。

次に、`s` が前の整数まででは作れない場合でも、`s - x` が作れていて、そこからまだ `x` を追加できるなら、`s` も作れます。

どちらでもないなら、`s` は作れません。

## 想定解法

```python
N, W = map(int, input().split())

a = []
m = []
for _ in range(N):
    x, c = map(int, input().split())
    a.append(x)
    m.append(c)

dp = [-1] * (W + 1)
dp[0] = 0

for x, c in zip(a, m):
    for s in range(W + 1):
        if dp[s] >= 0:
            # s は x を使わなくても作れる
            # したがって x は c 個すべて残っている
            dp[s] = c
        elif s >= x and dp[s - x] > 0:
            # s - x が作れて、そこから x を 1 個追加できる
            dp[s] = dp[s - x] - 1
        else:
            dp[s] = -1

print("Yes" if dp[W] >= 0 else "No")
```

時間計算量は `O(NW)`、空間計算量は `O(W)` です。

# 7. 共通して残せる文字列

これは最長共通部分列、LCS と呼ばれる問題です。

まず考え方として、2 次元 DP を考えると分かりやすいです。

```text
dp[i][j] = S の先頭 i 文字と T の先頭 j 文字の LCS の長さ
```

`S[i-1] == T[j-1]` なら、最後の文字を対応させることができます。

```text
dp[i][j] = dp[i-1][j-1] + 1
```

一方、最後の文字を対応させない場合は、どちらか片方の末尾を捨てることになります。

```text
dp[i][j] = max(dp[i-1][j], dp[i][j-1])
```

ただし、長さだけを求めればよいので、Python では 1 次元に圧縮できます。

## 想定解法

```python
S = input().strip()
T = input().strip()

# メモリを減らすため、T を短い方にする
if len(S) < len(T):
    S, T = T, S

M = len(T)
dp = [0] * (M + 1)

for ch in S:
    prev = 0
    for j in range(1, M + 1):
        old = dp[j]

        if ch == T[j - 1]:
            dp[j] = prev + 1
        else:
            dp[j] = max(dp[j], dp[j - 1])

        prev = old

print(dp[M])
```

時間計算量は `O(|S||T|)`、空間計算量は `O(min(|S|, |T|))` です。

# 8. 区間平均の最大化

数列を M 個の連続区間に分割します。

`dp[g][r]` を次のように定義します。

```text
先頭 r 個の要素を g 個の区間に分けたときの、平均値の総和の最大値
```

最後の区間が `a_l, a_{l+1}, ..., a_{r-1}` だとします。
このとき、先頭 `l` 個は `g - 1` 個の区間に分かれている必要があります。

したがって、遷移は次のようになります。

```text
dp[g][r] = max(dp[g-1][l] + average(l, r))
```

ここで `average(l, r)` は、0-indexed で `a[l], ..., a[r-1]` の平均です。

区間和を毎回 `O(区間長)` で求めると遅くなるので、累積和を使います。

```text
average(l, r) = (S[r] - S[l]) / (r - l)
```

これにより、各区間の平均を `O(1)` で計算できます。

## 想定解法

```python
N, M = map(int, input().split())
a = list(map(int, input().split()))

S = [0]
for x in a:
    S.append(S[-1] + x)

NEG = -10**100

prev = [NEG] * (N + 1)
prev[0] = 0.0

for g in range(1, M + 1):
    cur = [NEG] * (N + 1)

    for r in range(g, N + 1):
        best = NEG

        for l in range(g - 1, r):
            avg = (S[r] - S[l]) / (r - l)
            best = max(best, prev[l] + avg)

        cur[r] = best

    prev = cur

print(f"{prev[N]:.10f}")
```

時間計算量は `O(N^2M)`、空間計算量は `O(N)` です。
