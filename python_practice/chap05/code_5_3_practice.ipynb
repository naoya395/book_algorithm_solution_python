## 1. アルゴリズムとコードの詳しい解説

```python
INF = 1 << 60


def chmin(a: int, b: int) -> int:
    if a > b:
        return b
    return a


def main() -> None:
    n = int(input())
    h = list(map(int, input().split()))

    dp = [INF] * n
    dp[0] = 0

    for i in range(1, n):
        dp[i] = chmin(dp[i], dp[i - 1] + abs(h[i] - h[i - 1]))
        if i > 1:
            dp[i] = chmin(dp[i], dp[i - 2] + abs(h[i] - h[i - 2]))

    print(dp[n - 1])


if __name__ == "__main__":
    main()
```

### 目的

`chmin` を使って `dp[i]` を更新する、もらう DP の書き方です。

### 変数・関数の意味

- `dp[i]`: 足場 `i` への最小コストです。
- `chmin`: 現在値と候補値の小さい方を返します。
- `i`: 現在計算する足場です。

### 処理の流れ

1. 各 `i` について 1 歩前からの候補を必ず試し、`i > 1` なら 2 歩前からの候補も試します。

### 計算量

時間計算量は **O(n)** です。
