`func(i, w, a)` は、先頭 `i` 個の中からいくつか選んで合計 `w` を作れるかを返します。最後の要素を選ばない場合と選ぶ場合に分けます。


## 1. アルゴリズムとコードの詳しい解説

```python
def func(i: int, w: int, a: list[int]) -> bool:
    if i == 0:
        if w == 0:
            return True
        return False

    if func(i - 1, w, a):
        return True

    if func(i - 1, w - a[i - 1], a):
        return True

    return False


def main() -> None:
    n, w = map(int, input().split())
    a = list(map(int, input().split()))

    if func(n, w, a):
        print("Yes")
    else:
        print("No")


if __name__ == "__main__":
    main()
```

### 目的

`func(i, w, a)` は、先頭 `i` 個の中からいくつか選んで合計 `w` を作れるかを返します。最後の要素を選ばない場合と選ぶ場合に分けます。

### 変数・関数の意味

- `i`: 使える要素数です。
- `w`: 作りたい残りの和です。
- `a`: 候補の配列です。
- `func(i - 1, w, a)`: 最後の要素を選ばない場合です。
- `func(i - 1, w - a[i - 1], a)`: 最後の要素を選ぶ場合です。

### 処理の流れ

1. `i == 0` なら、`w == 0` のときだけ成功です。
2. 最後の要素を選ばない場合を試します。
3. 最後の要素を選ぶ場合を試します。
4. どちらかが成功すれば `True` を返します。

### 計算量

各要素について選ぶ／選ばないを試すため **O(2^n)** です。
