## 1. アルゴリズムとコードの詳しい解説

```python
def fibo(n: int) -> int:
    if n == 0:
        return 0
    if n == 1:

    return fibo(n - 1) + fibo(n - 2)
```

### 目的

`fibo(n)` は第 `n` 項のフィボナッチ数を返します。`n == 0`, `n == 1` を基本ケースにし、それ以外は `fibo(n - 1) + fibo(n - 2)` で求めます。

### 変数・関数の意味

- `n`: 求めたい項番号です。
- 基本ケース: `fibo(0)=0`, `fibo(1)=1` です。
- 再帰ケース: 直前 2 項の和を返します。

### 処理の流れ

1. `n == 0` なら 0 を返します。
2. `n == 1` なら 1 を返します。
3. それ以外は 2 つの再帰呼び出しの和を返します。

### 計算量

同じ値を何度も計算するため、この単純再帰は指数時間になります。
