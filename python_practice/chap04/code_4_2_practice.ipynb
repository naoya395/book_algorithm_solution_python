`func(n)` で和を計算しながら、呼び出し時と答えが戻る時にメッセージを表示します。再帰がどの順番で進み、戻ってくるかを観察するコードです。


## 1. アルゴリズムとコードの詳しい解説

```python
def func(n: int) -> int:
    print(f"func({n}) を呼び出しました")

    if n == 0:
        return 0

    result = n + func(n - 1)
    print(f"{n} までの和 = {result}")

    return result


def main() -> None:
    func(5)


if __name__ == "__main__":
    main()
```

### 目的

`func(n)` で和を計算しながら、呼び出し時と答えが戻る時にメッセージを表示します。再帰がどの順番で進み、戻ってくるかを観察するコードです。

### 変数・関数の意味

- `n`: 現在処理している値です。
- `result`: `n` までの和です。
- `func(n - 1)`: 1 つ小さい問題への再帰呼び出しです。

### 処理の流れ

1. 呼び出されたことを表示します。
2. `n == 0` なら 0 を返します。
3. `result = n + func(n - 1)` を計算します。
4. 計算結果を表示し、`result` を返します。

### 計算量

再帰の深さが `n` なので時間計算量は **O(n)** です。
