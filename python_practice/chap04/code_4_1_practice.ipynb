`func(n)` が `0 + 1 + ... + n` を返す再帰関数です。`n == 0` を終了条件にし、それ以外では `n + func(n - 1)` を返します。


## 1. アルゴリズムとコードの詳しい解説

```python
def func(n: int) -> int:
    if n == 0:
        return 0
    return n + func(n - 1)
```

### 目的

`func(n)` が `0 + 1 + ... + n` を返す再帰関数です。`n == 0` を終了条件にし、それ以外では `n + func(n - 1)` を返します。

### 変数・関数の意味

- `n`: 和を求めたい上限です。
- 終了条件: `n == 0` のとき 0 を返します。
- 再帰呼び出し: `func(n - 1)` で 1 つ小さい問題を解きます。

### 処理の流れ

1. `n == 0` なら 0 を返します。
2. そうでなければ `n` と `func(n - 1)` を足して返します。

### 計算量

再帰呼び出しが `n` 回程度なので時間計算量は **O(n)** です。
