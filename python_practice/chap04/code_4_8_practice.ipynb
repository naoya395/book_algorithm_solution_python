`memo` に計算済みのフィボナッチ数を保存し、同じ `n` を再計算しないようにします。単純再帰より大幅に高速になります。


## 1. アルゴリズムとコードの詳しい解説

```python
memo: list[int] = []


def fibo(n: int) -> int:
    if n == 0:
        return 0
    if n == 1:
        return 1

    if memo[n] != -1:
        return memo[n]

    memo[n] = fibo(n - 1) + fibo(n - 2)
    return memo[n]


def main() -> None:
    global memo
    memo = [-1] * 50

    fibo(49)

    for n in range(2, 50):
        print(f"{n} 項目: {memo[n]}")


if __name__ == "__main__":
    main()
```

### 目的

`memo` に計算済みのフィボナッチ数を保存し、同じ `n` を再計算しないようにします。単純再帰より大幅に高速になります。

### 変数・関数の意味

- `memo`: 未計算なら `-1`、計算済みなら答えを保存する配列です。
- `n`: 求めたい項番号です。
- `fibo(n - 1)`, `fibo(n - 2)`: 必要な小問題です。

### 処理の流れ

1. 基本ケースを返します。
2. `memo[n] != -1` なら保存済みの値を返します。
3. 未計算なら再帰で求めて `memo[n]` に保存します。
4. 最後に保存した値を返します。

### 計算量

各 `n` を高々 1 回計算するため **O(n)** です。
