配列 `f` に小さい項から順にフィボナッチ数を保存します。再帰ではなく反復で `f[n] = f[n - 1] + f[n - 2]` を計算します。


## 1. アルゴリズムとコードの詳しい解説

```python
def main() -> None:
    f = [0] * 50
    f[0] = 0
    f[1] = 1
    for n in range(2, 50):
        f[n] = f[n - 1] + f[n - 2]
        print(f"{n} 項目: {f[n]}")


if __name__ == "__main__":
    main()
```

### 目的

配列 `f` に小さい項から順にフィボナッチ数を保存します。再帰ではなく反復で `f[n] = f[n - 1] + f[n - 2]` を計算します。

### 変数・関数の意味

- `f`: フィボナッチ数を保存する配列です。
- `f[0]`, `f[1]`: 初期値です。
- `n`: 現在計算している項番号です。

### 処理の流れ

1. 長さ 50 の配列を用意します。
2. `f[0]=0`, `f[1]=1` を設定します。
3. `n=2` から順に `f[n]` を計算します。
4. 各項を表示します。

### 計算量

各項を 1 回ずつ計算するため **O(n)** です。
