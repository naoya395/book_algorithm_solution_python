`gcd(m, n)` は、`gcd(m, n) = gcd(n, m % n)` という性質を使って最大公約数を求めます。`n == 0` になったら `m` が答えです。


## 1. アルゴリズムとコードの詳しい解説

```python
def gcd(m: int, n: int) -> int:
    if n == 0:
        return m

    return gcd(n, m % n)


def main() -> None:
    print(gcd(51, 15))
    print(gcd(15, 51))


if __name__ == "__main__":
    main()
```

### 目的

`gcd(m, n)` は、`gcd(m, n) = gcd(n, m % n)` という性質を使って最大公約数を求めます。`n == 0` になったら `m` が答えです。

### 変数・関数の意味

- `m`, `n`: 最大公約数を求める 2 つの整数です。
- `m % n`: `m` を `n` で割った余りです。
- 終了条件: `n == 0` のとき `m` を返します。

### 処理の流れ

1. `n == 0` なら `m` を返します。
2. そうでなければ `gcd(n, m % n)` を返します。
3. 余りが 0 になるまで問題を小さくします。

### 計算量

ユークリッドの互除法は高速で、時間計算量は概ね **O(log min(m, n))** です。
